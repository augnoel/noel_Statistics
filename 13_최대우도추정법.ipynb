# 13. 최대우도추정법 Maximum Likelihood Estimation ; MLE

## 13.1 최대우도추정법의 기초
### 13.1.1 최대우도추정법의 정의
- 통계 모델에서 주어진 데이터에 가장 잘 맞는 모수parameter를 추정하는 방법
- 주어진 데이터가 관측될 확률을 가장 크게 만드는 모수를 찾는 방법
- 가장 큰 확률 = 최대 우도
- 수학적으로 주어진 데이터 X = (x_1, x_2, x_3 ... x_n)가 있을 때, 데이터가 어떤 모수 θ에 따라 생성
- 우도 함수 likelihood function, L(θ)
- 다음은 우도 함수를 최대화하는 θ값



### 13.1.2 우도 함수의 개념
- 주어진 데이터가 특정 모수 하에서 발생할 가능성을 나타냄
- 확률 밀도 함수PDF나 확률 질량 함수 PMF와 유사하지만, 중요한 차이점은 모수 변수로 간주
    - 확률 밀도 함수는 데이터를 입력을 받아 그 확률을 변환
- 우도 함수는 데이터를 고정시키고 모수를 변수로 취급
- 이 모수 값에서 이 데이터가 나올 확률이 얼머나 될지를 계산



- 우도 함수는 확률이 아닌 상대적인 가능성을 의미
- 특정 데이터를 생성할 가능성을 비교하는 데 사용
### 13.1.3 최대우도추정법의 직관적 이해
1. 관측된 데이터를 수집
    - 예시, 10번의 동전 던지기에서 7번이 앞면
2. 데이터가 특정 확률 모델을 따른다고 가정
    - 예시, 동전 던지기가 독립적이고 동일한 확률을 가지는 이항 분포를 따른다고 가정
3. 주어진 모델에서 데이터가 발생할 가능성을 나타내는 우도 함수 구성
    - 예시, 동전이 앞면이 나올 확률을 p라고 할때,

    

4. 우도 함수를 최대화하는 모수 p 찾기
    - 우도 함수 L(p)를 최대화하는 p는 데이터에서 관측된 빈도를 반영한 값
    - p = 0.7

## 13.2 최대우도추정의 계산
### 13.2.1 우도 함수의 설정
- 우도 함수를 변수로 취급하며, 데이터를 고정된 값으로 간주
- 우도 함수를 설정하는 단계
1. 모델 정의
    - 데이터를 설명할 수 있는 확률 모델 선택

    

2. 주어딘 모델을 바탕으로 우도 함수 설정



3. 여러 종류의 모델에 대해 설정 가능



### 13.2.2 로그 우도 함수 Log-Likelihood Function
- 계산이 복잡해 질 수 있는 우도 함수를 해결하기 위해 로그 우도 함수 사용



- 로그 함수는 단조 증가 함수이므로, 우도 함수를 최대화하는 모수 값은 로그 우도 함수도 최대화함
- MLE에서 로그 우도 함수의 최대화를 사용해도 동일한 추정치를 얻게 됨을 의미
- 로그 우도 함수는 곱셈 연산을 덧셈으로 변환해주므로 계산이 간단



### 13.2.3 우도 방정식과 최적화
- 로그 우도 함수를 최대화하려면 미분을 사용해 최대화되는 모수를 찾아야 함
- 이 과정을 우도 방정식 Likelihood Equation
1. 로그 우도 함수l(p) 미분
- 모수 p에 대해 미분



2. 우도 방정식
- 미분된 결과를 0으로 설정하여 우도 방정식을 만듬
- 이 식을 통해 최대 우도 추정치를 얻을 수 있음



3. 최적화
- 우도 방정식은 단순히 미분이 9이 지점을 찾는 것이므로 이 지점이 극대점인지 확인해야 함
- 이차 미분을 통해 극대점 확인하거나, 최적화 알고리즘을 사용해 직접 최대 우도 찾기

## 13.3 최대우도추정의 성질
### 13.3.1 일치성 Consistency
- 표본 크기가 증가함에 따라 최대우도추정치MLE는 참된 모수값에 수렴하는 성질
- 표본의 크기가 충분히 크면 MLE가 실제 모수 값과 거의 같아진다는 것을 의미
- 추정치가 일치성을 가지려면, 표본 크기n이 무한대로 갈 때, 추정치가 모수 θ에 수렴해야 함




- 일반적으로 MLE는 적절한 조건(독립적이고 동일한게 분포된 표본, 연속적일고 특정한 형태의 우도 함수)에서 일치성을 가짐
### 13.3.2 불편성 Unbiasedness
- 추정치의 기대값이 참된 모수값과 동일하다는 성질을 의미
- MLE가 평균적으로 참된 모수 값을 정확히 추정
- 추정치가 불편성을 가지려면, 그 기대값이 참된 모수 θ와 같아야 함



- MLE는 항상 불편한 것은 아님
- 일관성을 보장하지만, 일부 경우 불편 추정량이 될 수 있음
- 그러나, 표본 크기가 커지면 불편성의 정도가 줄어들어 MLE는 대체로 좋은 성질을 가짐
### 13.3.3 효율성 Efficiency
- 주어진 정보로부터 가능한 최소 분산을 가진 추정치를 의미
- 정보 이론에 서 가장 효율적인 의미
- 다른 불편 추정치들 중 MLE가 가장 작은 분산을 가질 수 있음을 나타냄
- 효율적 추정량은 불편 추정량 중 분산이 최소인 추정량
- 만약 MLE가 효율적이라면, 그 분산은 Cramer-Rao Lower Bound에 도달
- MLE는 적절한 조건 하에서 효율적
- 표본이 충분히 크고, 우도 함수가 특정 조건(정규성, 독립성, 동일한 분포)를 만족할 때 MLE가 효율적인 추정량이 될 수 있음

## 13.4 MLE의 응용
### 13.4.1 다양한 분포에서의 MLE
- 정규분포 Normal Distribution
    - 정규분포에서 모수는 평균 μ와 분산 σ^2
    - 주어진 표본에 대해 우도 함수를 최대화하여 MLE를 계산하면, 평균 μ의 MLE, 분산 σ^2의 MLE는 표본 분산이 됨
- 이항분포 Binomial Distribution
    - 이항분포에서 MLE는 성공 확률 p를 추정할 때 사용
    - 우도 함수는 주어진 성공 횟수와 표본 크기에 대해 정의
    - 이를 최대화하면 p의 MLE는 성공 횟수의 비율로 계산
- 포아송 분포 Poisson Distribution
    - 모수는 사건의 발생률 λ
    - 포아송 분포에서 MLE는 표본 평균으로부터 λ 추정
- 지수분포 Exponential Distribution
    - 지수분포의 모수는 발생률 λ 또는 평균 시간 간격
    - MLE는 주어진 표본의 역평균(1/표본평균)으로부터 λ 추정
### 13.4.2 회귀 분석에서의 MLE
- 선형 회귀 Linear Regression
    - 종속 변수 Y와 독립 변수 X 사이의 관계를 설명하는 모수 β를 추정
    - MLE는 잔차의 분포를 정규분포로 가정하고 이를 최대화하는 β값 추정
- 로지스틱 회귀 Logistic Regression
    - 이진 종속 변수에 대한 예측 모델
    - MLE는 로짓 함수(로지스틱 함수의 역함수)를 사용해 성공 확률을 추정하고 우도 함수를 최대화하여 회귀 계수 β 추정
### 13.4.3 MLE를 활용한 모델 적합
- 모델 적합도 평가
    - MLE는 모델의 적합도를 평가하는 중요한 방법 중 하나
    - 모델의 우도 함수를 최대화하는 모수 값을 찾는 것이 MLE의 핵심
    - 이를 통해 데이터에 가장 잘 맞는 모델 식별 가능
- 모델 비교
    - 여러 보델 간의 적합도를 비교할 때 MLE를 사용하여 각 보델의 우도 값을 계산 가능
    - 정보 기준(AIC, BIC) 등은 MLE를 기반으로 모델의 복잡성 적합도를 함께 고려하여 최적의 모델을 선택하는 데 사용
- 비선형 모델
    - 비선형 회귀 모델이나 혼합 효과 모델 등의 복잡한 모델에서도 MLE는 중요한 역할
    - 비선형 모델에서는 우도 함수를 직접 최대화하기 어려울 수 있으며, 이 때 수치적으로 최적화 기법을 사용하여 MLE 계산

## 13.5 MLE의 한계와 개선
### 13.5.1 극값 문제와 해의 존재성
- 극값 문제
    - MLE에서 우도 함수를 최대화하는 해를 찾기 위해서는 우도 함수의 극대값을 구해야 함
    - 그러나 우도 함수가 비선형이거나 복잡한 경우, 극대값이 여러개 존재할 수 있음
    - 이로 인해 MLE는 전역 극대값이 아닌 국소 극대값을 찾을 위험이 있음
- 해의 존재성
    - 특정 조건에서 우도 함수의 최대값을 구할 수 없는 경우도 있음
    - 표본이 지나치게 작거나, 데이터가 특이한 경우, 우도 함수가 수렴하지 않거나 해가 존재하지 않을 수 있음
    - 해의 존재성 문제는 MLE의 정확성에 큰 영향을 미치며, 이러한 상황에서는 다른 추정 방법이나 데이터의 특성에 맞는 모델링 접근이 필요
### 13.5.2 작은 표본 크기에서의 문제점
- 불안정한 추정
    - MLE는 표본 크기가 작을 때 불안정
    - 작은 표본 크기에서는 우도 함수가 제대로 정의되지 않거나, MLE가 큰 편향을 가질 수 있음
    - 결과적으로 추정값의 신뢰성이 떨어지게 됨
- 추정량 변동성
    - 표본 크기가 작을 때 MLE는 큰 변동성을 가질 수 있음
    - 동일한 데이터에 대해 MLE가 다른 결과를 도출할 가능성이 높아진다는 것을 의미하여 결과의 재현성을 낮춤
- 극단적인 값
    - 작은 표본 크기에서는 이상치나 극단적인 값이 MLE에 큰 영향을 줄 수 있음
    - 이러한 값들이 우도 함수에 큰 영향을 미쳐 잘못된 모수 추정을 초래
### 13.5.3 개선된 MLE 방법
- 정규화 Regularization
    - 모델의 복잡성을 제어하고 과적합을 방지하기 위한 방법
    - MLE에 정규화를 적용하면 우도 함수에 추가적인 제약 조건을 부여하여 모수의 크기나 복잡성 억제 가능
    - 라쏘(L1 정규화)와 릿지(L2 정규화)는 회귀 분석에서 정규화를 적용하는 대표적인 방법
    - 라쏘는 모수의 절대값을 최소화
    - 릿지는 모수의 제곱합을 최소화
- 패널티 함수 Penalty Function
    - 우동 함수에 추가적으로 포함되어 모수에 패널티 부여
    - MLE가 과도하게 복잡한 모델을 선택하지 않도록 유도
    - 모델의 일반화 능력을 향상
    - 패널티 함수를 추가함으로써 MLE는 복잡한 모델이 아닌 데이터 적합하면서도 단순한 모델을 선택할 수 있게 됨
- 베이지안 추정
    - MLE의 문제를 개선하기 위해 베이지안 추정을 사용하는 경우도 있음
    - 사전 분포를 통해 모수에 대한 추가 정보를 반영하며, 이는 작은 표본 크기나 극단적 데이터에 대해 더 안정적인 추정 가능

## 13.6 MLE와 다른 추정 방법 비교
### 13.6.1 최소제곱법Least Squares과의 비교
- 최소제곱법은 회귀 분석에서 오차의 제곱합을 최소화하여 회귀 계수를 추정
- 목표 함수
    - MLE는 주어진 데이터가 관찰될 가능성을 최대화 하는 모수값을 찾는 것이 목표
    - 우도 함수를 최대화하는 모수를 찾는 것이 MLE의 핵심
    - 최소제곱법은 오차의 제곱합을 최소화하는 회귀 계수를 찾는 것을 목표로 함
- 모델 가정
    - MLE는 데이터가 특정 분포(일반적으로 정규 분포)를 따른다는 가정하에 적용
    - 분포 가정에 따라 우도 함수가 정의되고, 최대화하는 모수 추정
    - 최소제곱법은 데이터의 오차가 독립적이고 동일한 분포(보통 정규 분포)를 따른다고 가정
    - 최소제곱법은 특히 오차가 정규 분포를 따를 때  MLE와 동일한 결과 도출
- 적용 범위
    - MLE는 다양한 분포와 모델에 적용
    - 최소제곱법은 주로 선형 회귀 분석에 적용, 오차가 정규 분포를 따르는 경우 유용
- 결과 해석
    - MLE는 추정된 모수가 해당 데이터에서 가장 가능성 있는 값을 나타냄
    - 확률론적으로 해석 가능
    - 최소제곱법은 오차의 최소화를 통한 추정을 목표, 기하학적인 해석 사용
### 13.6.2 베이즈 추정과의 비교
- 베이즈 추정
    - 사전 확률과 데이터를 결합하여 모수를 추정하는 방법
- 사전 정보 사용
    - MLE는 데이터를 기반으로 모수 추정, 사전 확률이나 사전 정보는 고려하지 않으며, 오로지 데이터의 우도만 최대화
    - 베이즈 추정은 사전 정보(사전 확률)를 반영하여 모수 추정, 특히, 데이터가 적거나 불확실성이 클 때 유리, 베이즈 추정에서 사후 확률 계산하여 모수 추정
- 확률적 해석
    - MLE 모수가 고정된 불확실서하지 않은 값이라고 가정하고 그 값을 추정, 모수에 대한 불확실성은 고려되지 않음
    - 베이즈 추정은 모수를 확률 변수로 간주하며, 모수에 대한 불확실성 반영, 베이즈 추정 결과는 모수의 분포로 표현되면 이는 사후 분포
- 계산 복잡성
    - MLE는 계산이 비교적 간단하며, 명확한 우도 함수를 가지는 경우 빠르게 결과를 도출
    - 베이즈 추정은 사전적 확률과 데이터의 결합을 필요로 하기 때문에 계산이 복잡, 고차원 문제에서는 계산량이 크게 증가 
- 실제 적용
    - MLE는 데이터가 충분히 크고 명확한 사전 정보가 없을 떄 효과적
    - 베이즈 추정은 사전 정보가 중요하거나 데이터가 적을 때, 추정에 불확실성을 반영할 떄 적합
### 13.6.3 모멘트 추정법과의 비교
- 모멘트 추정법
    - 데이터의 모멘트(평균, 분산)를 이요아여 모수를 추정하는 방법
- 추정 방식
    - MLE는 우도 함수를 최대화하는 모수를 찾음. 이 과정에서 데이터의 전체 분포를 고려하며 데이터가 최대한 관찰될 가능성을 중시
    - 모멘트 추정법은 데이터 모멘트를 계산하고 이를 이론적인 모멘트와 일치시키는 모수를 찾음. 예시로, 데이터의 평균을 이론적인 평균과 동일하게 만드는 모수 추정
- 계산의 단순성
    - MLE는 우도 함수를 설정하고 이를 최대화하는 과정이 필요. 이 과정이 복잡
    - 모멘트 추정법은 상대적으로 간단하며, 모멘트 계산을 통해 직접적으로 추정값을 도출. 계산이 빠르고 이해하기 쉬운 장점
- 효율성
    - MLE는 일반적으로 효율적이고 불편성을 가진 추정량 제공. 데이터가 충분히 크다면 MLE는 최소 분산을 가진 추정량 제공
    - 모멘트 추정법은 효율성 측면에서 MLE에 미치지 못할 수 있음. 표본 크기가 작거나 데이터의 분포가 복잡한 경우 모멘트 추정법은 MLE보다 큰 편향을 가질 수 있음
- 적용 범위
    - MLE는 다양한 분포와 상황에서 사용. 데이터의 확률 분포를 잘 반영
    - 모멘트 추정법은 주로 간단한 분포에 적용. 데이터의 모멘트가 중요한 경우에 유용. 데이터가 비정규적이거나 모멘트가 존재하지 않는 경우 사용 제한될 수 있음
