# 05 p-값과 통계적 유의성 p-value and Statistical Significance

## 5.1 p-값의 정의
### 1. p-값이란 무엇인가?
- 귀무가설이 참이라는 가정하에, 관찰된 데이터 또는 더 극단적인 데이터가 나타날 확률
- p-값은 0 ~ 1사이의 값을 가짐
- p-값이 작을수록 귀무가설이 참일 가능성이 낮아짐



### 2. p-값의 해석 방법
1. p < α
- 관찰된 데이터가 귀무가설을 기각할 충분한 증거를 제공한다고 해석
- 귀무가설을 기각하고, 대립가설 채택
2. p > α
- 관찰된 데이터가 귀무가설을 기각할 충분한 증거를 제공하지 못한다고 해석
- 귀무가설을 채택

### 중요한 점
- p-값은 귀무가설이 참일 확률이 아님
    - 귀무가설이 참이라는 가정 하에, 관찰된 데이터 또는 더 극단적인 데이터가 나타날 확률
- p-값이 작다고 해서 대립가설이 참이라는 것은 아님
    - 단지 귀무가설을 기각할 만한 증거가 있다는 것
- p-값은 단일 실험의 결과에 대한 확률적 해석을 제공
    - p-값만으로 결론을 내리는 것은 주의해야 함
    - 데이터의 맥락과 연구 설계를 고려한 해석이 필요

## 2. 유의수준Significance Level과 임계값
### 1. 유의수준 α의 역할
- 얼마만큼의 오류를 감수할 것인가를 결정하는 기준값
- 귀무가설이 참일 때, 이를 잘못 기각할 확률
- 제1종 오류가 발생할 확률
    - 귀무가설이 참이데도 불구하고 이를 기각하는 오류
- α는 귀무가설을 기각할지 말지를 결정하는 기준
- α가 낮을수록 검정이 더 엄격해지고, 제1종 오류를 범할 확률이 낮아짐
- p-값 < α, 귀무가설을 기각하고, 그렇지 않다면 기각하지 않음
### 2. 임계값과 p-값 비교
- 임계값 Critical Value
    - 유의수준 α에 대응하는 검정 통계량의 기준값
    - 검정 통계량이 임계값보다 극단적인 값에 위치하면, 귀무가설 기각
    - 임계값은 특정 분포(정규 분포, t-분포 등)에 따라 결정됨
- p-값
    - 실제로 관찰된 데이터에 대한 검정 통계량이 귀무가설 하에서 얼마나 극단적인지를 나타냄
- 귀무가설 기각 기준
    - 임계값 접근 : 검정 통계량이 임계값보다 극단적인 값이면 귀무가설을 기각
    - p-값 접근 : p-값 < α일 때 귀무가설 기각


## 3. p-값의 계산과 활용
### 1. p-값을 계산하는 방법
- 관찰된 데이터가 귀무가설 하에서 발생할 확률을 나타내는 값
1. 가설 설정
    - 귀무가설 H_0 : 관찰된 효과가 없다는 가정
    - 대립가설 H_1 : 관찰된 효과가 있다는 가정
2. 검정 통계량 계산
    - 관찰된 데이터를 기반으로 검정 통계량(z-값, F-값 등) 계산
3. 검정 통계량의 분포 확인
    - 검정 통계량이 어떤 분포(t-분포, 카이제곱 분포, 정규 분포)를 따르는지 확인
4. p-값 계산
    - 검정 통계량이 귀무가설 하에서 관찰된 데이터 보다 극단적인 값을 가질 확률
    - 검정 통계량의 분포에서 현재 관찰된 통계량 이상(또는 이하)의 값이 나타날 확률
### 2. p-값의 활용 예시
- 의학연구
- A/B테스트
- 심리학 연구

## 4. 통계적 유의성과 실제적 유의성의 차이
### 1. 통계적 유의성 vs. 실제적 유의성
- 통계적 유의성
    - 귀무가성를 기각할 수 있는지를 판단하는 기준
    - p-값을 사용하여 평가
    - p < α 일때 통계적으로 유의미하다고 함
        - p =0.03, α = 0.05 일때, 귀무가설을 기각, 관찰된 데이터가 우연에 의해 발생할 확률은 3% 로 매운 낮음을 의미
    - 이 결과가 우연히 발생할 확률이 얼마나 낮은가를 평가
- 실제적 유의성
    - 통계적으로 유의미한 결과가 실제 상황에서 중요한지 여부를 평가
    - 관찰된 효과의 크기나 그 효과가 실생활에서 의미가 있는지를 고려
    - 이 결과가 실생활에 영향을 미칠 정도록 중요한가를 평가
### 2. p-값에 대한 오해와 한계
- p-값은 효과의 크기를 알려주지 않는다
    - p-값이 작다는 것은 귀무가설이 맞을 확률이 낮다는 의미
    - 관찰된 효과의 크기가 아님
- p-값은 실험의 재현성을 보장하지 않음
    - p-값이 낮다고 해서 동일한 실험을 반복했을 때 항상 같은 결과를 얻는다는 보장은 없음
    - 표본 크기가 작거나 데이터에 노이즈가 많을 경우 p-값은 불안정할 수 있음
- p-값은 연구 설계의 품질을 반영하지 않음
    - p-값은 연구의 통계적 분석 결과를 나타내지만, 연구 설계나 데이터 수집 과정의 문제는 반영하지 않음
- p-값의 절단값 의존성
    - α = 0.05이고 p-값이 0.049, 0.051인 경우가 있을 때, 실제 두 결과는 크게 차이나지 않을 수 있지만, 하나는 유의미하고, 다른 하나는 유의미하지 않다고 결론지을 수 있음
- p-값의 대안
    - Effect Size
        - p-값 대신 Effect-Size를 함께 보고하면, 결과의 실제적 유의성을 더 잘 평가할 수 있음
    - 신뢰 구간 Confidence Interval 
        - p-값보다 더 많은 정보를 제공하며, Effect-Size와 불확실성을 동시에 나타냄
    - 베이지안 접근 
        - p-값이 의존하지 않고 데이터와 사전 확률을 결합하여 귀무가설과 대립가설에 대한 확률적 결론을 도출하는 방법

## 5. p-값의 보정 p-value Adjustment
- 다중 비교 문제에서 발생할 수 있는 오류를 줄이기 위한 통계적 방법
- 다중 비교
    - 여러 개의 가설 검정을 동시에 수행
### 1. 다중 비교 문제에서의 p-값 보정 방법
- 다중 비교를 수행할 때, 각 검정의 유의수준α를 고정하면, 검정을 반복할수록 제1종 오류의 확률이 증가
- 이는 여러 가설을 동시에 검정할 때, 개별 검정에서의 α를 그대로 적용하면 오류가 그대로 누적되기 때문
- 이를 방지하기 위해 p-값 보정이 필요
### 2. Bonferroni 보정, FDR 보정 등
- FWER ; Family-Wise Error Rate 조절
    - FWER은 한 번의 실험에서 적어도 하나의 잘못된 결론이 나올 확률을 의미
    - 이를 조절하기 위해 Bonferroni 보정과 같은 방법이 사용
- FDR ; False Discovery Rate 보정
    - FDR은 발견된 유의한 결과 중 실제로 유의하지 않은 비율을 의미
    - FDR는 FWER보다 덜 보수적
    - 여러 가설 검정에서 좀 더 유연하게 접근 
    - 해당 방법 붕 Benjamini-Hochberg 절차가 널리 사용
        - p-값에 대해 특정 기준을 적용하여 유의성 판단
        - p-값의 순위를 총 검정 수로 나눠 원하는 FDR 수준에 곱하여 사용
- Bonferroni 보정
    - 가장 간단한 다중 비교 보정 방법
    - 각 개별 검정의 유의수준을 전체 검정 수로 나누어 설정
    - 유의수준을 0.05로 설정하고 10개의 검정을 수행할 경우, 각 검정의 유의수준은 0.005가 됨
    - 매우 보수적이어서 실제로 유의한 결과를 놓칠 가능성이 있음

## 6. p-값에 대한 논쟁
### 1. p-값 사용의 비판과 대안
비판
- 오용과 오해
    - p값이 대립가설의 확증으로 잘못 이해하고 사용됨
- p-해킹
    - 원하는 결과를 얻기 위해 데이터나 분석 방법을 조작하는 경우
    - 연구의 신뢰성을 떨어뜨리고, 잘못된 과학적 발견을 초래
- 유의성 과대평가
    - p-값이 < 특정 임계값 일때 결과가 유의하다고 판단하는 방식은 결과의 실제 중요성을 과대평가할 수 있음
    - 과학적 발견의 재현성을 저해
대안
- 효과 크기와 신뢰 구간 
    - p-값 대신 Effect Size와 신뢰구간을 보고하여 결과의 실제적 중요성을 평가하는 방법
    - 결과의 크리와 방향을 더 명확히 전달할 수 있음
- 사전 확률과 베이지안 접근
    - 베이지안 통계는 사전 확률을 사용하여 결과를 해석
    - 데이터와 기존 지식을 결합하여 보다 직관적인 해석을 제공공
### 2. 베이지안 접근과 p-값 대체 지표
- 사전 확률 사용
    - 베이지안 통계는 사전 ㅘㄱ률을 사용하여 가설의 신뢰도를 평가
    - 기존 지식이나 전문가의 판단을 통합하여 보다 직관적이고 유연한 결과 해석을 가능하게 함
- 사후 확률 계산
    - 베인지안 방법은 데이터가 주어졌을 때 가설이 참일 확률을 계산
대표 지표
- 베이지안 신뢰 구간
    - 사후 확률 분포를 기반으로 하여 결과의 불확실성을 나타냄
- Bayes Factor
    - 두 가설의 상대적 지지 정도를 나타냄
    - p-값보다 더 많은 정보를 제공
