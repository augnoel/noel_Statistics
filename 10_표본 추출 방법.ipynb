# 10 표본 추출 방법

## 10.1 표본 추출의 기초
### 10.1.1 표본 추출의 정의
- 표본 추출Sampling
    - 모집단 전체를 조사하지 않고 그 중에서 일부를 선택하여 조사하는 과정
    - 모집단poulation은 연구자가 관심을 가지는 모든 대상을 의미
    - 표본은 이 모집단에서 선택된 일부의 대상
- 표본 추출의 목적
    - 표본을 통해 모집단의 특성을 추정하거나 모집단에 대한 결론을 내리는 것
    - 전체 모집단을 조사하는 것보다 시간과 비용이 절약되며, 실용적인 연구 방법으로 널리 사용
### 10.1.2 표본 추출의 중요성
- 표본 추출은 연구 결과의 신뢰성과 타당성에 직접적인 영향을 미침
- 잘 설계된 표본 추출은 모집단을 대표할 수 있는 표본을 확보하게 해줌
- 연구 결과를 모집단 전체로 일반화 가능
- 만약 표본 추출인 적절치 않다면, 조사 결과가 편향될 수 있어 잘못된 결론으로 이어질 수 있음
### 10.1.3 표본 추출 과정
1. 모집단 정의
- 연구자가 관심을 가지는 전체 집단을 정의
2. 표본 프레임 결정
- 표본 프레임 sampling frame
    - 모집단에 포함된 모든 구성원들의 목록 도는 DB를 작성하거나 접근 가능한 자료를 확보
- 가능한 모집단을 완벽하게 반영해야 함
3. 표본 추출 방법 선택
- 연구의 목적과 모집단의 특성을 고려하여 적절한 표본 추출 방법 선택
4. 표본 크기 결정
- 표본의 크기는 연구의 신뢰성과 관련 있음
- 표본이 클수록 모집단을 정확히 반영할 가능성이 높아짐
- 하지만 시간과 비용도 증가
- 통계적 방법을 이용해 적절한 표본 크기 결정
5. 표본 추출
- 선택한 방법에 따라 실제 표본을 추출
6. 표본 검토 및 수정
- 추출된 표본이 모집단을 대표할 수 있는지 검토
- 피룡하다면 보정이나 표본 추출 수행

## 10.2 확률 표본 추출 Probability Sampling
- 모집단의 모든 구성원이 표본으로 선택될 확률이 동일하게 보장되는 표본 추출 방법
- 연구의 대표성과 신뢰성을 높이기 위한 방법
- 각 구성원이 표본으로 선택될 확률이 명확히 정의됨
### 10.2.1 단순 랜덤 표본 추출 Simple Random Sampling
- 가장 기본적인 확률 표본 추출 방법
- 모집단의 모든 구성원이 동일한 확률로 선택될 기회를 가짐
- 표본이 무작위로 추출됨
- 과정
    1. 모집단 모든 구성원에게 번호 부여
    2. 무작위로 번호를 추출하거나, 난수 생성기를 사용하여 특정 수의 표본 선택
- 장점
    - 표본이 모집단을 잘 대포할 가능성이 높으며, 통계적 분석이 간단
    - 추출과정이 간단하고 명확
- 단점
    - 모집단이 매우 클 경우, 실용적이지 않을 수 있음
    - 표본 프레임이 정확하지 않으면 문제 발생할 수 있음
### 10.2.2 층화 표본 추출 Stratified Sampling
- 모집단을 서로 다른 층으로 나누고 각 층에서 단순 랜덤 표본 추출을 수행하는 방법
- 각 층은 공통된 특성을 가진 하위 집단으로 구성됨
- 층 내에서 단순 랜덤 표본 추출이 이뤄짐
- 과정
    1. 모집단을 의미있는 기준에 따라 여러 층으로 나눔
    2. 각 층에서 일정한 비율로 단순 랜덤 표본 추출
- 장점
    - 각 층의 특성을 정확히 반영할 수 있어 모집단의 이질성을 잘 다룰 수 있음
    - 층 간 비교 가능
- 단점
    - 층을 나누는 기준이 적절하지 않으면 표본이 왜곡될 수 있음
    - 복잡합 모집단에서 층화 설정이 어려움
### 10.2.3 군집 표본 추출 Cluster Sampling
- 모집단을 몇개의 군집으로 나눈 후, 일부 군집을 무작위로 선택하여 그 군집 내의 구성원을 조사하는 방법
- 군집은 모집단의 축소판으로 간주
- 각 군집은 모집단을 대표할 수 있어야 함
- 과정
    1. 모집단을 여러 개의 군집으로 나눔
    2. 군집 중 일부를 무작위로 선택
    3. 선택된 군집 내 모든 구성원 조사
- 장점
    - 표본 추출 과정이 간단하고 비용이 절감됨
    - 넓은 지리적 범위에서 효율적으로 표본 추출 가능
- 단점
    - 군집 간 동질성이 낮을 경우 표본이 모집단을 잘 대표하지 못할 수 있음
    - 한 군집에 과도하게 의존할 경우 결과가 왜곡될 위험이 있음
### 10.2.4 계통 표본 추출 Systematic Sampling
- 일정 간격에 따라 표본 추출하는 방법
- 모집단의 크기가 크거나 명부가 정렬되어 있을 때 유용
과정
    1. 모집단 전체 크기N와 필요한 표본 크기n 결정
    2. N/n의 값을 계산하여 이를 기준으로 표본의 추출할 간격k 설정
    3. 첫 번째 표본은 무작위로 선택하고 이후 간격 k만큼 떨어진 구성원을 표본으로 선택
- 장점
    - 표본 추출이 매우 간단하고 시간과 비용이 절약
    - 명부가 정렬된 경우 특정 패턴을 쉽게 반영 가능
- 단점
    - 모집단이 어떤 주기성을 띤 경우 표본이 왜곡될 수 있음
    - 첫 번째 표본 선택이 무작위로 이뤄지지 않으면 표본이 비대표적일 수 있음

## 10.3 비확률 표본 추출 Non-Probability Sampling
- 모집단의 구성원들이 표본으로 선택될 확률이 일정하지 않으며, 무작위성이 보장되지 않는 표본 추출 방법
- 연구자가 편리하게 접근할 수 있는 구성원들이나 특정 기준을 만족하는 구성원들을 주로 선택
- 비확률 표본 추출은 연구의 목표에 따라 유용할 수 있지만, 대표성 부족으로 인해 일반화가 어려움
### 10.3.1 편의 표본 추출 Convenience Sampling
- 연구자가 접근하기 쉬운 구성원들을 표본으로 선택
- 신간과 비용이 절약되며 연구 초기 탐색적 연구에 유용
- 장점
    - 빠르고 간편하게 표본 수집 가능
    - 제한된 시간, 비용으로 연구 수행 가능
- 단점
    - 표본이 모집단을 대표하지 않을 가능성이 큼
    - 연구 결과를 일반화하기 어려움
### 10.3.2 판단 표본 추출 Judgmental Sampling
- 연구자가 특정 기주에 따라 모집단에서 표본을 선택하는 방법
- 연구 목적에 맞게 '적합'하다고 판단되는 구성원들을 선택
- 연구자의 전문성과 판단이 표본 선택에 중요
- 특정 특성을 가진 집단을 연구할 때 사용
- 장점
    - 특정 연구 목적에 맞게 표본 구성 가능
    - 연구자가 필요로 하는 특정 정보나 특성을 가진 구성원들을 직전 선택 가능
- 단점 
    - 주관적인 판단에 의해 표본이 선택되므로 편향이 발생할 수 있음
    - 일반화 가능성이 매우 낮음
### 10.3.3 할당 표본 추출 Quota Sampling
- 모집단의 특정 하위 집단에 따라 할당된 비율에 맞춰 표본을 선택하는 방법
- 층화 표본 추출과 비슷하지만, 각 하위 집단 내에서 무작위로 표본을 선택하지 않는 점에서 차이가 있음
- 모집단의 하위 집단으로 나눈 후, 각 집단에서 특정 비율로 표본 추출
- 할당된 베율에 따라 각 집단에서 표본 채움
- 장점
    - 특정 집단 간의 비교를 쉽게 가능
    - 다양한 하위 집단을 연구에 포함 가능
- 단점
    - 각 하위 집단 내 표본이 무작위로 선택되지 않기 때문에 대표성이 떨어질 수 있음
    - 할당 기준이 적절하지 않으면 연구 결과에 편향에 발생할 수 있음
### 10.3.4 눈덩이 표본 추출 Snowball Sampling
- 초기 몇몇 구성원들을 통해 그들의 추천을 받아 표본 확장해 나가는 방법
- 주로 희귀하거나 접근하기 어려운 모집단을 연구할 때 사용
- 초기 표본이 다른 사람들을 추천하고 추천된 사람들 역시 새로운 표본을 추천하는 방식으로 표본 확대
- 네트워크를 통해 표본이 확장되는 방식
- 장점
    - 접근하기 어려운 모집단에 효과적으로 접근 가능
    - 네트워크를 통해 표본을 신속히 확대
- 단점
    - 초기 표본의 선택에 따라 표본의 편향이 심해질 수 있음
    - 표본이 특정 네트워크에 한정될 수 있어 대표성이 떨어질 수 있음

## 10.4 표본 크기 결정
- 연구의 신뢰성과 타당성 확보를 위해 매운 중요
- 적절한 표본 크기 선택은 연구 결과의 일반화 가능성을 높이고, 통계적 검정의 힘을 증대시킬 수 있음
### 10.4.1 표본 크기의 중요성
- 표본 크기는 연구의 결과에 크게 영향을 미침
- 너무 작은 표본은 통계적 유의성을 발견 못할 수 있음
- 너무 큰 표본은 자원 낭비와 비효율성을 초래할 수 있음
- 신뢰성 향상
    - 적절한 표본 크기는 연구 결과의 신뢰성을 높임
    - 더 큰 표본은 모집단의 특성을 더 정확하게 반영할 가능성이 있음
- 통계적 검정력
    - 표본 크기기 클수록 통계적 검정력이 증가
    - 효과가 실제로 존재할 때 이를 발견할 확률이 높아짐을 의미
- 오류 감소
    - 작은 표본 크기는 무자위 오류를 증가시킬 수 있어 이로 인해 잘못된 결론에 도달할 위험 증가
### 10.4.2 표본 크기 계산 방법
- 연구의 목표, 사용된 통계적 검정, 기대되는 효과 크기, 유의 수준, 통계적 검정력1-β에 따라 달라짐
- 기본 적인 표본 크기 계산 공식
    - 평균 비교를 위한 표본 크기 계산

    

    - 프로그램 활용
        - 연구에서 사용하는 통계 패키지나 전용 프로그램을 통해 표본 크기 계산 가능
        - 다양한 연구 설계와 가정하에 표본 크기 계산에 유용
    - 경험적 접근
        - 비슷한 연구에서 사용된 표본 크기를 참조하거나 파일럿 연구를 통해 대략적인 표본 크기르 추정 가능
### 10.4.3 표본 크기와 신뢰 수준
- 신뢰 수준은 연구 결과가 참값을 포함할 확률을 의미
- 표본 크기는 신뢰 구간의 폭에 영향
- 신뢰 수준의 영향
    - 신뢰 수준이 높을수록 신뢰 구간이 넓어짐
    - 신뢰 수준이 높아질수록 더 큰 표본 크기 필요
- 표본 크기와 신뢰 구간
    - 표본 크기가 커지면 신뢰 구간의 폭이 좁아져 추정의 정확성이 높아짐
    - 모집단의 모수에 대해 더 정확한 추정을 가능케 함
- 실질적인 의미
    - 신뢰 수준이 너무 높거나 낮으면 실질적인 해석이 어려움

## 10.5 표본 추출의 실제 적용
### 10.5.1 표본 추출에서의 오류
- 표본 추출 과정에서 발생하는 오류는 연구 결과의 신뢰성과 타당성을 저하시킴
- 표본 오차 Sampling Error
    - 모집단에서 일부를 추출하는 과정에서 발생하는 자연스러운 변동
    - 표본이 모집단을 완벽하게 대표하지 못하기 때문에 발생
    - 표본 크기를 증가시키면 표본 오차 줄일 수 있임
    - 확률 표본 추출 방법을 통해 표본이 모집단을 더 잘 대표할 수 있음
- 비표본 오차 Non-Sampling Error
    - 표본 추출 외의 다른 요인으로 인해 발생하는 오류
    - 데이터 수집, 처리, 분석 과정에서 발생
    - 응답 오류 Response Error
        - 응답자가 질문에 부정확하게 답변하는 경우 발생
    - 무응답 오류 Non-Response Error
        - 선택된 표본이 응답을 하지 않을 때 발생
    - 측정 오류 Measurement Error
        - 잘못된 도구나 방법으로 데이터를 측정할 때 발생
- 이를 낮추기 위해 설문지의 명확한 설계, 충분한 응답률 확보, 정확한 데이터 수집 절차를 통해 비표본 오차 낮출 수 잇음
### 10.5.2 대표성 있는 표본 확보
- 확률 표본 추출 방법 사용
    - 단순 랜덤 표본 푸출, 층화 표본 추출, 군집 표본 추출 등 확률 표본 추출 방법을 사용하면, 모든 모집단 구성원이 표본으로 선택될 확률이 같아짐
    - 표본의 대표성이 높일 수 있음
- 층화 표본 추출
    - 모집단이 이질적인 그룹으로 구성된 경우, 각 그룹에서 적절한 비율로 표본을 추출하면 더 대표성 있는 표본 확보 가능
- 충분한 표본 크기
    - 표본 크기가 충분히 크면 모집단의 특성을 더 잘 반영할 수 있음
    - 대신, 너무 큰 표본은 자원 낭비를 초래할 수 있으며로 적절한 크기 유지
- 비표본 오차의 최소화
    - 데이터 수집 과정에서 비표본 오차를 줄이는 것도 표본 대표성을 높일 수 있음
    - 명확한 설문지 설계, 훈련된 조사자를 활용하여 오류 낮출 수 잇음
### 10.5.3 다양한 상황에서의 표본 추출 전략
- 시간과 자원이 제한된 경우
    - 편의 표본 추출 Convenience Sampling
        - 접근 가능한 대상들로부터 표본을 추출하는 방법 사용
        - 대표성이 낮아질 수 있어 결과를 일반화할 때 주의
- 모집단이 이질적인 경우
    - 층화 표본 추출 Stratified Sampling
        - 모집단이 여러 하위 그룹으로 구성되어 있고, 각 그룹이 상이한 특성을 가질 경우 사용
        - 각 그룹에서 적절한 비율로 표본을 추출함으로써 대표성 높일 수 있음
- 지리적으로 넓은 모집단
    - 군집 표본 추출 Cluster Sampling
        - 지리적 구역이나 그룹을 먼저 선정한 후, 일부를 선택하여 표본 추출하는 것이 효율적
- 특정 특성을 가진 소수 집단
    - 눈덩이 표본 추출 Snowball Sampling
        - 특정 특성을 가진 소수 집단을 대상으로 연구할 때, 초기 참가자에게 다른 참가자를 소개받는 방식으로 표본 확대
