# 2. 확률 분포 Probability Distribution
- 정의
    - 확률 변수가 가질 수 있는 값들과 그 값들이 나타날 확률의 분포
    - 확률 질량 함수 PMF와 확률 밀도 함수 PDF로 표현

## 2.1 이산 확률 분포 Discrete Probability Distribution
- 정의
    - 확률 변수가 취할 수 있는 값이 유한하거나 셀 수 있는 무한 개수인 분포
    - 확률 변수가 정수와 같은 분리된 값을 가질 때 이를 이산 확률 변수
    - 이 변수의 분포를 이산 확률 분포
    - 확률 변수는 셀 수 있는 개수의 값
    - 각 값에 대해 확률이 정의되며. 이 확률의 총합은 1
    - 확률 질량 함수PMF를 통해 각 값에 대한 확률을 나타냄

### 2.1.1 이항 분포 Binomial Distribution
- 정의
    - 고정된 횟수의 독립 시행에서 특정 사건이 발생하는 횟수를 나타내는 이산 확률 분포
    - 각 시행은 성공 또는 실패로 나뉨
    - 각 시행의 성공 확률은 동일
    - 시행 횟수 n : 실험 또는 시행의 총 횟수
    - 성공 확률 p : 각 시행에서 성공할 확률
    - 실패 확률 q = 1 - p : 각 시행에서 실패할 확률
    - 확률 변수 X는 성공 횟수, X는 0 ~ n 까지의 정수 값을 가짐
- 확률 질량 함수 (PMF)
    
    

- 평균과 분산
    - 평균 : μ=np
    - 분산 : σ^2 = np(1−p)
- 예시 및 파이썬 코드

```python
# 한글 폰트
from matplotlib import font_manager, rc

font_path = "C:/Windows/Fonts/malgun.ttf"  # Windows의 경우 "Malgun Gothic" 폰트 경로
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# 이항 분포의 매개변수
# 시행 횟수
n = 10  
# 성공 확률
p = 0.5  

# 이항 분포를 따르는 확률 변수
X = np.arange(0, n+1)
binom_pmf = stats.binom.pmf(X, n, p)

# 이항 분포의 평균과 분산
mean = n * p
variance = n * p * (1 - p)

# 결과 출력
print(f"이항 분포의 평균: {mean}")
print(f"이항 분포의 분산: {variance}")

# PMF 시각화
plt.bar(X, binom_pmf, color='skyblue')
plt.xlabel('성공 횟수 (k)')
plt.ylabel('확률 P(X=k)')
plt.title(f'이항 분포 (n={n}, p={p})')
plt.show()
```

### 2.1.2 포아송 분포 Poisson Distribution
- 정의
    - 일정한 시간 또는 공간 내에서 사건이 발생하는 횟수를 나타내는 이산 확률 분포
    - 사건이 독립적으로 발생하고, 사건의 발생률이 일정한 상황에서 사용
    - 큰 시행 횟수와 작은 성공 확률을 가진 이항 분포의 극한으로 간주할 수 있음
    - 단위 시간 또는 공간 내에서 사건 발생 횟수를 모델링
    - 사건은 독립적으로 발생
    - 단위 시간 또는 공간 애세어 사건 발생률이 일정
- 확률 질량 함수 (PMF)

    

- 평균과 분산
    - 평균 : μ=λ
    - 분산 : σ^2 = λ
- 예시 및 파이썬 코드

```python
# 포아송 분포의 매개변수
lambda_ = 4  # 평균 사건 발생 횟수

# 포아송 분포를 따르는 확률 변수
X = np.arange(0, 15)  # 0부터 14까지의 사건 발생 횟수
poisson_pmf = stats.poisson.pmf(X, lambda_)

# 포아송 분포의 평균과 분산
mean = lambda_
variance = lambda_

# 결과 출력
print(f"포아송 분포의 평균: {mean}")
print(f"포아송 분포의 분산: {variance}")

# PMF 시각화
plt.bar(X, poisson_pmf, color='lightgreen')
plt.xlabel('사건 발생 횟수 (k)')
plt.ylabel('확률 P(X=k)')
plt.title(f'포아송 분포 (λ={lambda_})')
plt.show()
```

### 2.1.3 기하 분포 Geometric Distribution
- 정의
    - 일련의 독립적 시행에서 첫 번째 성공이 나타나기까지 실패한 횟수
    - 각 시행에서 성공할 확률은 동일
    - 성공 확률 : p
    - 실패 확률 q = 1 - p
    - 독립적 시행
        - 각 시행은 다른 시행의 결과에 영향을 받지 않음
    - 첫 번째 성공까지의 실패 횟수를 모델링
- 확률 질량 함수 (PMF)

    

- 평균과 분산
    - 평균 : μ = 1 / p
    - 분산 : σ^2 = (1 - p) / p^2
- 예시 및 파이썬 코드

```python
# 기하 분포의 매개변수
p = 0.5  # 성공 확률

# 기하 분포를 따르는 확률 변수
X = np.arange(1, 11)  # 첫 번째 성공까지의 실패 횟수 (1부터 시작)
geom_pmf = stats.geom.pmf(X, p)

# 기하 분포의 평균과 분산
mean = 1 / p
variance = (1 - p) / (p ** 2)

# 결과 출력
print(f"기하 분포의 평균: {mean}")
print(f"기하 분포의 분산: {variance}")

# PMF 시각화
plt.bar(X, geom_pmf, color='lightcoral')
plt.xlabel('실패 횟수 (k)')
plt.ylabel('확률 P(X=k)')
plt.title(f'기하 분포 (p={p})')
plt.show()
```

### 2.1.4 초기하 분포 Hypergeometric Distribution
- 정의
    - 유한한 모집단에서 비복원 추출을 통해 특정 개수의 성공을 얻을 확률을 나타내는 분포
    - 복원하지 않고 추출하는 경우를 다룸
    - 모집단 크기 N : 모집단 전체 크기
    - 성공 항목의 수 K : 모집단에서 성공으로 간주하는 항목의 수
    - 추출 횟수 n : 모집단에서 추출하는 횟수
    - 성공 횟수 k : 추출한 항목 중 성공 항목의 수
- 확률 질량 함수 (PMF)

    

- 평균과 분산
    - 평균 : μ = n * K / N
    - 분산

        
​
- 예시 및 파이썬 코드

```python
# 초기하 분포의 매개변수
N = 52  # 모집단 크기 (전체 카드 수)
K = 4   # 성공 항목의 수 (에이스 카드 수)
n = 10  # 추출 횟수 (뽑는 카드 수)

# 초기하 분포를 따르는 확률 변수
X = np.arange(0, min(K, n) + 1)  # 가능한 성공 횟수
hypergeom_pmf = stats.hypergeom.pmf(X, N, K, n)

# 초기하 분포의 평균과 분산
mean = n * (K / N)
variance = n * (K / N) * ((N - K) / N) * ((N - n) / (N - 1))

# 결과 출력
print(f"초기하 분포의 평균: {mean}")
print(f"초기하 분포의 분산: {variance}")

# PMF 시각화
plt.bar(X, hypergeom_pmf, color='lightblue')
plt.xlabel('성공 횟수 (k)')
plt.ylabel('확률 P(X=k)')
plt.title(f'초기하 분포 (N={N}, K={K}, n={n})')
plt.show()
```

## 2.2 연속 확률 분포
- 확률 변수가 연속적인 값을 가질 때 확률 변수가 특정 값이 아닌 범위 내에 있을 확률을 계산
- 확률 밀도 함수를 사용하여 특정 구간에 속할 확률을 계산
- 확률 밀도 함수 PDF : 특정 값이 아닌 구간에 대한 확률을 나타내는 함수
- 누적 분포 함수 CDF : 확률 변수가 특정 값 이하일 때 확률을 나타내는 함수
- 특정 값의 확률은 0, 확률은 구간으로 계산

### 2.2.1정규 분포 Normal Distribution
- 정의
    - 정규 분포는 연속 확률 분포에서 가장 중요한 분포
    - 자연현상과 사회현상에서 많은 변수들이 따르는 분포
    - 데이터가 중앙값 주위에 대칭적으로 분포
    - 평균과 표준편차로 정의
    - 평균 μ : 데이터의 중앙값
    - 표준편차 σ : 데이터 분포 정도를 나타내는 값
- 확률 밀도 함수 (PDF)



- 평균과 분산
    - 평균 μ
    - 분산 σ^2
- 정규 분포의 특성 (68-95-99.7 규칙)
    - 정규 분포에서 데이터의 68%,  95%, 99.7%가 각각 평균으로 부터 1, 2, 3 표준편차 내에 존재하는 특성
    - 68% : 데이터의 68%가 평균 ± 1 표준편차 내에 존재
    - 95% : 데이터 95%가 평균 ± 2 표준편차 내에 존대
    - 99.7% : 데이터의 99.7%가 평균 ± 3 표준편차 내에 존재
- 표준 정규 분포와 Z-변환
    - 표준 정규 분포는 평균이 0, 표준편차가 1인 정규 분포
    - 다른 정규 분포를 표준 정규 분포로 변환 가능하며 이를 Z - 변환
- 예시 및 파이썬 코드

```python
# 정규 분포의 매개변수
mu = 70  # 평균
sigma = 10  # 표준편차

# 정규 분포를 따르는 확률 변수
X = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)  # 평균 ± 4표준편차 범위
normal_pdf = stats.norm.pdf(X, mu, sigma)

# 정규 분포의 평균과 분산
mean = mu
variance = sigma**2

# 결과 출력
print(f"정규 분포의 평균: {mean}")
print(f"정규 분포의 분산: {variance}")

# PDF 시각화
plt.plot(X, normal_pdf, color='blue')
plt.xlabel('값 (X)')
plt.ylabel('확률 밀도 함수 (PDF)')
plt.title(f'정규 분포 (μ={mu}, σ={sigma})')
plt.show()
```

### 2.2.2 균등 분포 Uniform Distribution
- 정의
    - 특정 구간 내의 모든 값이 동일한 확률로 나타나는 분포
    - 연속 균등 분포에서는 확룰 변수가 일정 구감 [a, b] 내에서 균일하게 분포
    - a : 구간의 시작점 (최소값)
    - b : 구간의 끝점
    - 구간  [a. b]내에서 모든 값이 동일한 확률로 나타남
- 확률 밀도 함수 (PDF)



- 평균과 분산
    
    

- 예시 및 파이썬 코드

```python
# 균등 분포의 매개변수
a = 1  # 구간의 시작점
b = 5  # 구간의 끝점

# 균등 분포를 따르는 확률 변수
X = np.linspace(a, b, 1000)
uniform_pdf = stats.uniform.pdf(X, a, b - a)

# 균등 분포의 평균과 분산
mean = (a + b) / 2
variance = ((b - a) ** 2) / 12

# 결과 출력
print(f"균등 분포의 평균: {mean}")
print(f"균등 분포의 분산: {variance}")

# PDF 시각화
plt.plot(X, uniform_pdf, color='green')
plt.xlabel('값 (X)')
plt.ylabel('확률 밀도 함수 (PDF)')
plt.title(f'균등 분포 (a={a}, b={b})')
plt.ylim(0, 1/(b - a) * 1.1)  # y축 범위 설정
plt.show()
```

### 2.2.3지수 분포 Exponential Distribution
- 정의
    - 사건 간의 시간 간격을 모델링하는 데 사용되는 분포
    - 주로 포아송 과정에서 사건이 발생하 때까지의 대기 시간을 나타냄
    - 일정한 비율로 사건이 발생하는 과정에서 두 사건 간 시간을 모델링
    - λ : 사건 발생의 평균률. 단위 시간당 사건이 발생하는 평균 횟수
    - 지수 분포는 '기억 없음' 속성을 가지고 있어 과거 사건이 미래 사건에 영향을 미치지 않음
- 확률 밀도 함수 (PDF)
    - x가 증가할수록 확률은 기하급수적으로 감소
    - 시간이 지날수록 사건이 발생할 확률이 줄어드는 패턴

    

- 평균과 분산
    - 평균 : μ = 1/λ
        - 사건이 발생하는 평균 시간 간격
    - 분산 σ^2 = 1/λ^2
        - 시간 간격의 변동성
        - 분산이 클수록 간격의 변동이 큼
        - 분산이 작을수록 변동이 적음
- 예시 및 파이썬 코드

```python
# 지수 분포의 매개변수
# 사건 발생의 평균률
lambda_param = 0.2 

# 지수 분포를 따르는 확률 변수
X = np.linspace(0, 30, 1000)
exponential_pdf = stats.expon.pdf(X, scale=1/lambda_param)

# 지수 분포의 평균과 분산
mean = 1 / lambda_param
variance = 1 / (lambda_param ** 2)

# 결과 출력
print(f"지수 분포의 평균: {mean}")
print(f"지수 분포의 분산: {variance}")

# PDF 시각화
plt.plot(X, exponential_pdf, color='purple')
plt.xlabel('시간 (X)')
plt.ylabel('확률 밀도 함수 (PDF)')
plt.title(f'지수 분포 (λ={lambda_param})')
plt.show()
```


### 2.2.4감마 분포 Gamma Distribution
- 정의
    - 주로 사건의 총 대기 시간을 모델링하는 데 사용
    - 두개의 매개변수인 형상 파라미터 α와 비율 파라미터 β를 가짐
    - 다양한 형태의 분포를 생성
    - 포아송 분포와 지수 분포의 일반화된 형태
    - 형상 파라미터 k : 감마 분포의 형태를 결정하는 매개변수
        - k가 정수일 때, 감마 분포는 k개의 독립적인 지수 분포의 합과 동일
    - 스케일 파라미터 θ(비율 파라비터 β) : 스케일을 조절하는 매개변수
        - 사건 발생 간격의 평균을 조절
        - θ =  1/λ
        - λ : 지수 분포에서 비율 파라미터
- 확률 밀도 함수 (PDF)

    

- 평균과 분산
    - 평균 μ=kθ : 기대되는 사건 발생 시간의 평균 간격
    - 분산 σ^2 = kθ^2 : 사건 발생 간격의 변동성
- 예시 및 파이썬 코드

```python
# 감마 분포의 매개변수
# 형상 파라미터
alpha = 2
# 비율 파라미터 
beta = 3 

# 감마 분포를 따르는 확률 변수
X = np.linspace(0, 20, 1000)
gamma_pdf = stats.gamma.pdf(X, a=alpha, scale=beta)

# 감마 분포의 평균과 분산
mean = alpha * beta
variance = alpha * (beta ** 2)

# 결과 출력
print(f"감마 분포의 평균: {mean}")
print(f"감마 분포의 분산: {variance}")

# PDF 시각화
plt.plot(X, gamma_pdf, color='orange')
plt.xlabel('시간 (X)')
plt.ylabel('확률 밀도 함수 (PDF)')
plt.title(f'감마 분포 (α={alpha}, β={beta})')
plt.show()
```

### 2.2.5 베타 분포 Beta Distribution
- 정의
    - 확률 변수 값을 0과 1 사이의 구간에서 모델링하는 데 사용
    - 베이지안 통계에서 사전 분포로 사용
    - 확률이나 비율을 모델링할 때 유용
    - 두 개의 매개 변수인 형상 파라미터 α, β에 의해 정의
    - 형상 파라미터 α, β : 분포 모양을 결정, 각 구간에서 비율을 조절
- 확률 밀도 함수 (PDF)

    

- 평균과 분산

    

- 예시 및 파이썬 코드

```python
# 베타 분포의 매개변수
alpha = 2  # 형상 파라미터
beta = 5   # 형상 파라미터

# 베타 분포를 따르는 확률 변수
X = np.linspace(0, 1, 1000)
beta_pdf = stats.beta.pdf(X, alpha, beta)

# 베타 분포의 평균과 분산
mean = alpha / (alpha + beta)
variance = (alpha * beta) / ((alpha + beta)**2 * (alpha + beta + 1))

# 결과 출력
print(f"베타 분포의 평균: {mean}")
print(f"베타 분포의 분산: {variance}")

# PDF 시각화
plt.plot(X, beta_pdf, color='green')
plt.xlabel('확률 변수 (X)')
plt.ylabel('확률 밀도 함수 (PDF)')
plt.title(f'베타 분포 (α={alpha}, β={beta})')
plt.show()
```

### 2.2.6 카이제곱 분포 Chi-Square Distribution
- 정의
    - 가설 검정, 독립성 검정 및 적합도 검정에서 사용
    - 독립적인 표준 정규 분포 변수의 제곱 합으로 정의
    - 자유도 Degree of Freedom ; df
        - 카이제곱 분포는 자유도에 의해 결정
        - 변수의 개수와 관련
        - 일반적으로 표본의 수에서 제약 조건의 수를 뺀 값으로 계산
- 확률 밀도 함수 (PDF)

    

- 자유도와 분산
    - 자유도 df
        - 주어진 통계적 검정의 설계에 따라 결정
        - 독립성 검정에서 자유도는 (r−1)×(c−1)로 계산
        - r과 c는 행렬의 행, 열의 수
    
    

- 예시 및 파이썬 코드

```python
# 카이제곱 분포의 자유도
df = 4  # 자유도

# 카이제곱 분포를 따르는 확률 변수
X = np.linspace(0, 20, 1000)
chi2_pdf = stats.chi2.pdf(X, df)

# 카이제곱 분포의 평균과 분산
mean = df
variance = 2 * df

# 결과 출력
print(f"카이제곱 분포의 평균: {mean}")
print(f"카이제곱 분포의 분산: {variance}")

# PDF 시각화
plt.plot(X, chi2_pdf, color='blue')
plt.xlabel('값 (X)')
plt.ylabel('확률 밀도 함수 (PDF)')
plt.title(f'카이제곱 분포 (df={df})')
plt.show()
```
