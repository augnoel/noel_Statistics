# 11. 중심 극한 정리 Central Limit Theorem ; CLT

## 11.1 중심 극한 정리의 정의
### 11.1.1 중심 극한 정리란 무엇인가?
- 통계학과 확률론에서 중요한 정리
- 매우 큰 표본에서 표본 평균의 분포가 정규분포에 가까워지는 현상
- 모집단의 분포가 어떤 형태이든 상관없이, 충분히 큰 표본을 반복해서 추출하여 표본 평균을 계산하면 그 평균들의 분포가 정규분포에 가까운진다는 의미
- 모집단이 정규분포를 따르지 않더라도, 표본의 크기가 커질수록 표본 평균의 분포는 평균이 μ, 분산이 σ^2/n인 정규분포N에 근접(μ는 모집단 평균, σ^2는 모집단 분산, n은 표본 크기)
1. 정규분포 근사
- 모집단이 정규분포가 아니더라도, 표본 평균의 분포는 표본 크기가 커질수록 정규분포에 가까워짐
2. 추론 통계 기반
- 표본 평균을 사용하여 모집단의 평균을 추정할 때, 중심 극한 정리를 통해 신뢰 구간을 설정하거나 가설 검정 수행 가능
### 11.1.2 중심 극한 정리의 수학적 표현
- 모집단 평균 μ와 분산 σ^2를 가진 임의의 확률 분포에서 표본 크기 n의 표본을 반복적으로 추출한다면, 각 표본에서 얻은 표본 평균X_n일때, 중심 극한 정리에 따르면 n이 충분히 클 때, 다음이 성립



- 표본의 크기가 클수록 분모가 작아져, X_n이 μ근처에 밀집
- 표본 평균의 분포는 모집단 분포의 형태와 상관없이 정규분포에 수렴

## 11.2 중심 극한 정리의 조건
### 11.2.1 독립성 가정
- 중심 극한 정리가 성립하기 위한 기본적인 전제 조건
- 표본 내의 관측값들이 서로 독립적
- 하나의 관측값이 다른 관측값에 영향을 주지 않아야 한다는 것
- 독립성 가정의 중요한 예외
    - 복원 추출 With Replacement Sampling
        - 각 표본을 추출 후 다시 모집단에 반환하는 경우, 추출된 표본들이 독립성을 유지 가능
    - 큰 모집단
        - 매운 큰 모집단에서 비복원 추출하는 경우, 비록 엄밀한 독립성이 없더라도 중심 극한 정리는 유효할 수 있음
### 11.2.2 동일한 분포 가정
- 각 표본이 동일한 확률 분포에서 추출되었다는 것
- 모든 관측값이 동일한 확률 분포를 따르며, 각각의 분포가 평균 μ와 분산 σ^2를 가지는 상황을 가정
- 표본 평균을 계산할 때, 각 관측값이 동일한 기대값과 분산을 가지는 것이 필요함
- 만약 표본들이 서로 다른 분포에서 추출되었다면, 표본 평균의 분포가 정규분포에 수렴하지 않을 수 있음
### 11.2.3 표본 크기의 영향
- 표본의 크기가 충분히 커야만 중심 극한 정리가 유효하게 작용
- 표본 평균의 분포가 정규분포에 수렴
- 작은 표본 크기
    - 표본 평균의 분포는 모집단 분포의 영향을 더 많이 받음
    - 정규분포로 근사되지 않을 수 있음
    - 비대칭 분포나 꼬리가 두꺼운 분포일숡 표본 크기의 영향일 증가
- 큰 표본 크기
    - 표본의 분포는 정규분포에 가까워짐
    - 일반적으로 표본 크기가 30이상이면 대부분의 경구 중심 극한 정리가 적용될 수 있음
    - 분포가 극단적일 경우 더 큰 표본 크기가 필요할 수 있음

## 11.3 중심 극한 정리의 중요성
- 다양한 표본 분포들이 특정 조건 하에서 정규분포에 근사될 수 있다는 강력한 이론적 근거
### 11.3.1 표본 분포와 정규분포
- 표본 분포 Sampling Distribution
    - 동일한 모집단에서 여러 표본을 추출하여 계산된 통계량의 분포
    - 중심 극한 정리는 이 표본 분포가 표본 크기가 충분히 클 때 정규분포에 가까워짐을 보장
- 정규 본포의 수렴
    - 중심 극한 정리는 모집단의 분포가 무엇이든지 간에 표본 크기가 충분히 크다면 표본 평균의 분포가 평균 μ와 표본 오차σ/n^1/2를 가지는 정규분포로 수렴한다고 설명
    - 모집단의 분포가 비정규적이거나 알 수 없더라도 중심 극한 정리는 표본 평균이 정규분포에 근사하므로 통계적 추론을 가능케 함
### 11.3.2 추론 통계에서의 역할
- 추론 통계의 기초
    - 추론 통계는 표본 데이터를 바탕으로 모집단의 특성을 추정하고, 가설 검정을 통해 의사 결정을 내리른 과정
    - 중심 극한 정리는 이러한 추론 통계의 기초를 제공
- 모집단에 대한 추정
    - 중심 극한 정리 덕분에 우리는 표본 평균을 이용해 모집단의 평균을 추정 가능
    - 표본 평균이 정규 분포에 근사하므로, 신뢰구간 계산이나 가설 검정에서 정규분포의 성질을 활용 가능
- 가설 검정
    - 중심 극한 정리는 표본 통계량이 정규분포를 따른다는 가정 하에, 귀무가설이 참일 때 통계량의 분포를 예측 가능
    - 이로 인해 p-값을 계산하고, 통계적 유의성 평가 가능
- 신뢰구간 설정
    - 신뢰구간은 모집단의 모수를 포함할 확률이 높은 구간을 의미
    - 중심 극한 정리를 통해 표본 평균의 분포가 정규분포에 근사함을 알고 있기 때문에, 신뢰구간을 설정할 때 정규분포 활용 가능

## 11.4 중심 극한 정리의 응용
### 11.4.1 모집단 분포가 비정규적인 경우
- 비정규 분포와 중심 극한 정리
    - 모집단이 정규분포를 따르지 않는 경우에도 중심 극한 정리를 통해 표본 평균이 정규분포로 근사될 수 있음
    - 표본 크기가 충분히 클 때, 표본 평균의 분포가 모집단의 분포 형태에 상관없이 정규분포에 가까워짐
- 표본 크기의 중요성
    - 표본 크기가 작을 때 표본 평균의 분포가 정규분포에 근사하지 않을 수 있음
    - 표본 크기를 증가시키면 비정규적인 모집단 분포에서도 표본 평균의 분포는 점점 정규분포에 근사
    - 모집답이 비정규적일지라도 정규분포 기반의 통계적 추론 가능
### 11.4.2 중심 극한 정리와 신뢰 구간
- 신뢰 구간의 설정
    - 중심 극한 정리 덕분에 우리는 표본 평균이 정규분포를 따른다고 가정
    - 이를 바탕으로 신뢰 구간 계산
    - 모집단 평균에 대해 불확실성을 표현 가능
- 신뢰 구간 계산
    - 표본 평균, 표준 오차, 그리고 정규분포의 적절한 분위수 사용
    - 중심 극한 정리로 인해 표본 평균이 정규분포를 따르므로, 우리는 신뢰수준을 기준으로 신뢰 구간 설정 가능
### 11.4.3 중심 극한 정리와 가설 검정
- 가설 검정에서의 역할
    - 귀무가설이 참이라는 가정하에 표본 데이터를 분석하여 귀무가설을 기각할지 여부를 결정
    - 중심 극한 정리는 이 과정에서 표본 평균이 분포를 정규분포로 근사할 수 있게 하여 p-값 계산과 임계값 설정에 중요
- p-값과 정규분포
    - 중심 극한 정리에 따르면, 표본 평균의 분포는 정규분포로 근사되므로, p-값을 계산할 때 정규분포 사용 가능
    - 정규분포의 특성을 활용하여 귀무가설이 참일 때 실제 관찰된 표본 통계량보다 극단적인 확률을 계산하는 것

## 11.5 중심 극한 정리의 시뮬레이션
### 11.5.1 컴퓨터 시뮬레이션을 통한 이해
- 시뮬레이션의 목적
    - 중심 극한 정리를 직관적으로 이해하기 위해 다양한 모집단 분포에서 샘플을 추출하고 그 샘플 평균이 어떻게 분포하는지 시뮬레이션을 통해 관찰
    - 특히 비정규적인 분포를 가진 모집단에서 표본 크기가 커질수록 샘플 평균의 분포가 어떻게 정규분포에 가까워지느지 확인
- 시뮬레이션 절차
1. 모집단 정의
- 모집단의 분포를 정의
- 모집단으 정규분포일 수도 있고 , 비정규분포(이항분포, 포아송분포, 균등분포, 기하분포 등)일 수 있음
2. 표본 추출
- 모집단에서 여러 번 표본을 무작위로 추출
- 표본 크기는 다양한 값을 사용 가능
- 표본 크기가 결과에 미치는 영향 관찰 가능
3. 표본 평균 계산
- 각 표본에서 평균을 계산
- 이 과정을 여러 번 반복하며 얻어진 표본 평균들은 데이터셋을 형성
4. 표본 평균의 분포 관찰
- 추출된 모든 표본 평균들의 분포를 시각화
- 이 분포가 중심 극한 정리에 따라 정규분포에 가까워지는지 확인
