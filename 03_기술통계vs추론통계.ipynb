# 3. 기술 통계 vs 추론 통계
## 3.1 기술 통계 Descriptive Statistics
### 3.1.1 기술 통계의 정의
- 수집된 데이터를 요약하고 설명하는 데 사용되는 통계 기법
- 데이터를 이해하기 쉽게 변환하는 데 중점
- 데이터를 시각적으로 표현하거나, 데이터를 간단한 숫자로 요약하는 것을 목표
- 데이터를 분석하기 위한 첫번째 단계
- 데이터를 파악하고 더 
- 복잡한 분석(추론 통계)으로 넘어가기 전 전체적인 패턴을 이해하는 데 도움
### 3.1.2 중심 경향 측정 Central Tendency
- 데이터 세트의 중심 또는 대표값을 나타내는 통계적 지표
- 이 측정값은 데이터를 요약하고, 특정 집단의 일반적인 특성을 이해하는 데 중요한 역할
1. 평균 Mean
- 모든 데이터 값의 합을 데이터 개수로 나눈 값
- 데이터의 중심을 수치적으로 나타냄
- 모든 데이터 값이 동일한 가중치
- 이상치에 민감
    - 매우 크거나 작은 값이 포함될 경우 평균이 크게 변함



2. 중앙값 Median
- 데이터를 크기 순으로 정렬했을 때 중앙에 위치한 값
- 데이터가 짝수 개일 경우, 중앙에 위치한 두 값의 평균을 중앙값으로 사용
- 이상값이 덜 민감, 데이터 세트에 극단적인 값이 포함되어 있어도 중앙값은 큰 영향 받지 않음
- 데이터틔 분포가 비대칭적일 때, 중앙값이 더 적합한 중심 경향 측정값이 될 수 있음

3. 최빈값 Mode
- 데이터 세트에서 가장 자주 나타나는 값
- 하나 이상일 수 있음
- 모든 데이터 값이 동일한 빈도를 가지면 최빈값 없음
- 범주형 데이터에서 특히 유용
- 데이터에 여러 개의 최빈값이 있을 수 있음
    - 다봉 분포Multimodal Distribution

### 3.1.3 산포 측정
- 산포 Dispersion
    - 데이터가 평균이나 중앙값 주변에서 얼마나 퍼져 있는지 나타내는 통계적 지표
    - 산포 측정값을 통해 데이터가 얼마나 일관된지, 얼마나 변동성이 큰지 이해
1. 범위 Range
- 데이터 세트에서 가장 큰 값과 가장 작은 값의 차이
- 이는 데이터 전체의 산포를 간단히 나타내는 방법
- 계산이 간단하고 데이터 산포 정도를 빠르게 파악 가능
- 이상값에 민감. 데이터에 극단적인 값이 포함되어 있을 경우 범위가 크게 변함
- 데이터의 전체 분포보다는 최댓값과 최솟값에만 의존하기 때문에 전체적일 산포를 정확히 반영하지 못할 수 있음



2. 분산 Variance
- 데이터 값들이 평균으로부터 얼마나 떨어져 있느지를 제곱하여 평균한 값
- 데이터의 산포 정도를 수치적으로 나타내는 지표
- 음수 값이 없고, 이상값의 영향을 크게 받음
- 데이터 분포를 더 정확히 반영, 표준편차 제곱과 동일



3. 표준편차 Standard Deviation
- 분산의 제곱근
- 데이터 값들이 평균으로부터 얼마나 떨어져 있는지 나타냄
- 분산의 단위를 데이터와 동일하기 만들어 해석을 더 쉽게 함



### 3.1.4 분포의 형태
1. 왜도 Skewness
- 데이터 분포의 비대칭 정도를 나타내는 지표
- 분포가 좌우로 얼마나 기울어져 있는지 설명
- 왜도의 절대값이 클수록 비대칭 정도가 큼
- 양의 왜도 Positive Skewness 
    - 분포의 꼬리가 오른쪽으로 길게 늘어진 형태
    - 평균 > 중앙값
    - 데이터의 대부분이 분포의 왼쪽에 몰려있음
- 음의 왜도 Negative Skewness
    - 분포의 꼬리가 왼쪽으로 길게 늘어진 형태
    - 평균 < 중앙값
    - 데이터 대부분이 분포의 오른쪽에 몰려있음
- 왜도가 0에 가까운 경우 
    - 분포가 대칭임
    - 정규 분포의 왜도가 0



2. 첨도 Kurtosis
- 데이터 분포의 꼬리가 두꺼운 정도와 분포의 중심이 얼마나 뾰족한지를 나타내는 지표
- 양의 첨도 Leptokurtic
    - 분포의 중심이 더 뾰족하고 꼬리가 더 두꺼운 형태
    - 정규분포보다 극단적인 값이 많음
-음의 첨도 Platykurtic
    - 분포의 중심이 더 평평하고 꼬리가 더 얇은 형태
    - 정규분포보다 극단적인 값이 적음
- 첨도가 0인 경우
    - 정규분포와 동일한 수준의 뾰족함과 꼬리 두께



### 3.1.5 상관 관계와 공분산 Correlation and Covariance
1. 공분산 Covariance
- 두 변수 관계를 측정하는 지표
- 한 변수가 증가할 때 다른 변수가 증가하거나 감소하는 경향
- 공분산의 값은 변수의 단위에 의존하기 때문에, 값의 크기만으로는 두 변수 간의 관계의 강도를 비교하기 어려움
- 공분산 > 0 : X, Y는 양의 상관 관계
- 공분산 < 0 : X, Y는 음의 상관 관계
- 공분사 = 0 : X, Y가 선형 관계가 없음



2. 상관 관계 Correlation
- 두 변수 간의 선형 관계
- 공분산을 각 변수의 표준편차로 나누어 정규화한 값
- 상관 계수는 -1 ~ 1 사이의 값을 가짐
- 상관 계수가 1에 가까울수록, 강한 양의 선형 관계
- 상관 계수가 -1에 가까울수록, 강한 음의 선형 관계
- 상관 계수 = 0 : 두 변수 간에 선형 관계가 없음

### 3.1.6 데이터 시각화 기법
1. 히스토그램 Histogram
- 연속형 데이터의 분포를 시각화
- 일정한 간격으로 나누어 구간에 해당하는 데이터의 빈도를 막대그래프로 나타냄

```python
import matplotlib.pyplot as plt

# 예시 데이터
data = [12, 15, 17, 19, 22, 22, 24, 25, 28, 30, 32, 35, 38, 40]

# 히스토그램 그리기
plt.hist(data, bins=5, color='skyblue', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram Example')
plt.show()
```

2. 박스 플롯 Box Plot
- 데이터 분포, 중심, 변동성을 시각적으로 나타내는 방법
- 데이터의 사분위수를 강조하여 이상값 확인에 유용

```python
# 예시 데이터
data = [12, 15, 17, 19, 22, 22, 24, 25, 28, 30, 32, 35, 38, 40]

# 박스 플롯 그리기
plt.boxplot(data)
plt.title('Box Plot Example')
plt.ylabel('Value')
plt.show()
```

3. 산점도 Scatter Plot
- 두 변수 간의 관계를 시각화할 때 사용
- 각 데이터포인트는 x, y좌표를 가짐
- 두 변수간 관계를 파악

```python
# 예시 데이터
x = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y = [1, 4, 9, 16, 25, 36, 49, 64, 81]

# 산점도 그리기
plt.scatter(x, y, color='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot Example')
plt.show()
```

4. 막대 그래프 Bar Chart
- 범주형 데이터를 시각화하는데 사용
- 각 범주에 대한 데이터 값을 막대의 길이로 표현

```python
# 예시 데이터
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 8, 4]

# 막대 차트 그리기
plt.bar(categories, values, color='green')
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Bar Chart Example')
plt.show()
```

5. 파이 차트 Pie Chart
- 데이터의 구성 비율을 시각화하는 데 사용
- 전체 데이터를 원형으로 나누어 각 부분이 전체에서 차지하는 비율

```python
# 예시 데이터
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]

# 파이 차트 그리기
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart Example')
plt.show()
```

## 3.2 추론 통계 Inferential Statistics

### 3.2.1 추론 통계의 정의
- 모집단 전체에 대해 결론을 내리기 위해 표본 데이터를 분석하고 해석하는 통계 방법론
- 표본으로부터 얻은 정보를 바탕으로 모집단의 특성을 추정하거나 가설을 검정
- 표본이 아닌 전체 모집단에 대한 결론을 도출
- 모집단 파라미터 추정
    - 모집단의 평균, 비율, 분산 등과 같은 파라미터를 포본 데이터를 통해 추정
    - 점 추정 point estimation, 구간 추정 interval estimation
- 가설 검정 
    - 주어진 가설에 대한 증거를 제공하기 위해 표본 데이터를 사용
    - 가설이 모집단 전체에 타당한지 여부 판단
- 모집단 특성 예측
    - 표본 데이터를 기반으로 모집단의 미래 결과나 특성 예측
- 추론 통계와 기술 통계의 차이
    - 기술 총계
        - 표본 데이터나 모집단 데이터를 요약하고 정리하는데 초점
        - 평균, 중앙값, 표준편차 등을 계산하여 데이터의 중심 경향과 변동성 설명
    - 추론 통계 
        - 기술 통계로 부터 얻은 표본 데이터를 사용하여 모집단 전체에 대해 추론하거나 예측을 수행
        - 표본의 분석 결과를 바탕으로 모집단의 특성에 대해 결론내리는 과정
     
### 3.2.2 표본과 모집단의 개념
1. 모집단 Population
- 연구나 조사 대상이 되는 전체 집합
- 특정 특성을 공유하는 모든 개체나 관측값의 집합
- 모집단은 보통 너무 크기 때문에 전체를 조사하는 것이 비현실적이거나 불가능할 수 있음
- 모집단데 대한 통계적 추정을 위해 표본 데이터를 사용
2. 표본 Sample
- 모집단에서 선택된 부분 집합
- 모집단의 특성을 추정하기 위해 사용
- 표본은 모집단을 대표
- 그 결과를 바탕으로 모집단의 특성을 일반화
- 모집단의 특성을 추정하기 위해 통계적 방법을 사용하여 분석
- 표본 크기와 선정 방법에 따라 표본의 신뢰성 및 정확도가 달라짐
3. 표본과 모집단의 관계
- 표본 추출
    - 모집단에서 표본을 선택하는 과정
    - 표본은 무작위 추출되거나 다양한 방법(단순 무작위 추출, 계통적 추출, 층화 추출 등 )
- 표본 통계량
    - 표본에서 계산된 통계적 측정값
    - 모집단의 모수를 추정하는데 사용
    - 표본 평균, 표본 분산 등
- 모수 Parameter
    - 모집단의 특성을 나타내는 값
    - 모집단 평균, 모집단 분산 등
- 표본과 모집단의 분석 과정
1. 표본 설계 및 추출
    - 모집단의 특성을 반영하는 표본을 설계하고 무작위로 추출
    - 표본의 크기와 선정 방법은 연구의 목적에 맞게 조정
2. 표본 분석
    - 표본 데이터를 수집하여 평균, 분산, 비율 등 다양한 통계적 측정값을 계산
    - 데이터의 중심 경향, 변동성, 분포 등을 분석
3. 모집단 특성 추정
    - 표본 분석 결과를 바탕으로 모집단의 모수 추정
    - 추정치는 점 추정, 구간 추정 방식으로 제공
4. 가설 검정 
    - 표본 데이터를 사용하여 특정 가설 검정
    - 모집단의 특성에 대한 결론을 도출하거나 가설 검증

### 3.2.3 표본 통계량과 모수 추정 
1. 표본 통계량 Sqmple Statistic
- 표본 데이터에서 계산된 값
- 모집단의 모수를 추정하는 데 사용
- 표본 평균 Sample Mean, x
    - 표본 데이터의 평균값
    - 모집단의 평균을 추정하는 데 사용

    

- 표본 분산 Sample Variance, s^2
    - 표본 데이터의 분산
    - 모집단이 분산을 추정하는 데 사용

    

- 표본 표준편차 Sample Standard Deviation,s
    - 표본 분산의 제곱근
    - 데이터의 변동성

    

- 표본 비율 Sample proportion
    - 이항 분포에서 성공의 비율
    - 모집단 비율을 추정하는 데 사용

    

2. 모수 추정
- 표본 통계량을 사용하여 모집단의 모수를 추정하는 과정
- 점 추정 Point Estimation
    - 표본 통계량을 사용하여 모집단의 모수를 단일 값으로 추정하는 방법
    - 표본 평균을 사용하여 모집단 평균을 추정
- 구간 추청 Interval Estimation
    - 표본 통계량을 사용하여 모집단 모수가 포함될 가능성이 높은 범위를 추정하는 방법
    - 주로 신뢰구간을 통해 표현
    - 95% 신뢰구간을 사용하여 모집단 평균이 특정 구간에 포함될 확률 95%임을 나타냄
    - 모집단 평균의 신뢰구간

        

### 3.2.4 신뢰 구간 Confidence Intervals ; CI
- 표본 데이터를 바탕으로 모집단의 모수를 추정할 때, 그 모수가 포함될 범위를 제공하는 구간
- 신뢰 구간은 추정값의 불확실성을 정량화하며, 추정값이 특정 범위에 포함될 확률을 명시
- 주어진 신뢰수준에 따라 표본에서 추정된 모수가 포함될 수 있는 범위를 나타냄
    - 예 : 95% 신뢰구간은 모수가 이 범위 안에 있을 확률이 95%임을 의미



1. 신뢰 구간 계산
- 모수가 정규 분포를 따를 때와 그렇지 않을 때로 나뉨
- 정규 분포를 가정한 경우

    

- 모집단 분산이 알려지지 않은 경우



- 비율에 대한 신뢰구간



### 3.2.5 가설 검정 Hypothesis Testing
- 주어진 데이터에 기반하여 가설의 진위를 평가하는 통계적 방법
- 모집단의 특성에 대한 주장을 검증하는 데 사용
- 연구자가 제시한 가설이 데이터에 의해 지지되는지 판단
1. 가설 Hypothesis
- 모집단의 모수에 대한 추정이나 주장
- 귀무가설 Null Hypothesis, H0
    - 기본적인 주장
    - 변화를 주장하지 않거나, 두 그룹 간의 차이가 없다고 가정
- 대립가설 Alternative Hypothesis, H_a
    - 귀무가설과 반대되는 주장
    - 변화나 차이가 있다고 가정
2. 가설 검정 과정
    1. 가설 설정
    2. 검정 통게량 계산
        - 데이터에서 특정 통계량을 계산하여 귀무가설이 참일 때 분포를 기반으로 검정 통계량 계산
    3. 유의 수준 설정
        - α = 0.05 또는 0.01 설정
        - 귀무가설을 기각할 임계값의 확률
    4. p-value 계산
        - 귀무가설이 참일 때, 현재 또는 더 극단적인 데이터를 얻을 확률
    5. 결론 도출
        - p-value < α : 귀무가설을 기각하고 대립가설 채택
        - p-value > α : 귀무가설 기각 불가. 데이터가 귀무가설을 지지
3. 가설 검정 유형
- One Sample t-Test
    - 표본 평균이 모집단 평균과 유의미하게 다른지 검정

     

- Two Sample t-Test
    - 두 그룹의 평균이 유의미하게 다른지 검정

    

- Chi-Squared Test
    - 범주형 데이터에서 관측된 빈도와 기대 빈도 간의 차이를 검정

    

- ANOVA ; Analysis of Variance
    - 3개 이상의 그룹 간 평균 차이를 검정

    

### 3.2.6 p-value와 유의성
1. p-value
- 주어진 귀무가설이 참이라는 전제하에, 현재 관측된 데이터나 그 보다 더 극단적인 결과가 나타날 확률
- p-value가 작아질수록 관측된 데이터가 귀무가설을 지지하지 않는다는 증거가 강해짐
2. 유의 수준 Significance α
- 귀무가설을 긱할 기준이 되는 확률값
- 연구가자 미리 설정하는 임계값
- α = 0.05 : 연구자가 5%의 확률로 귀무가설을 잘못 기각할 위험을 감수하겠다는 의미
3. p-value의 해석
- p-value ≤ α : 귀무가설 기각, 대립가설 채택. 관측된 결과는 우연에 의한 것이라 보기 어렵다는 결론
- p-value > α : 귀무가설 채택, 정확하게는 데이터가 귀무가설을 지지한다고 할 수 없지만, 귀무가설을 기각할 만한 증거가 없음을 이미
4. p-value와 유의성의 관계
- p-value는 유의성을 판단하는 핵심 기준
- p-vlaue가 작다고 반드시 실제로 중요한 결과를 의미하즌 것은 아님
- 통계적 유의성 : p-value ≤ α 일때, 귀무가설을 기가함으로써 대립가설 지지
- 실질적 중요성 :  연구결과가 통계적으로 유의미하더라도, 그 결과가 실질적으로 중요한지 평가해야 함

### 3.2.7 오류 유형 Type of Errors
1. 제1종 오류 Type I Error
- 실제로는 참인 귀무가설을 기각하는 오류
- 즉 귀무가설이 참임에도 불구하고 대립가설을 채택하는 오류
- 제1종 오류의 확률
    - 연구자가 설정한 α이 제1종 오류가 발생할 최대 확률
2. 제2종 오류 Type II Error
    - 실제로는 거짓인 귀무가설을 기각하지 못하는 오류
    - 즉 대립가설이 참임에도 불구하고, 귀무가설을 채택하게 되는 오류
    - 제2종 오류의 확률 β
        - β값이 클수록 제2종 오류를 범할 가능성이 큼
        - β이 작을수록 통계 검정의 검정력Power, 즉 실제로 대립가설이 참일 때 이를 올바르게 검출할 확률이 커짐
        - 검정력 = 1 - β
3. 오류 유형의 상관관계
- 제1종 오류와 제2종 오류 사이에는 상관관계가 있음
- α를 낮추면 제1종 오류 가능성을 줄일 수 있지만, 그 대신 제2종 오류의 가능성β가 커질 수 있음
- 유의 수준과 검정력의 트레이드오프
    - 유의 수준을 낮춰 제1종 오류를 줄이면, 제2종 오류가 발생할 가능성이 높아져, 검정력이 낮아짐
    - 연구 설계에서는 이러한 트레이드오프를 고려하여 적절한 유의 수준과 검정력을 설정해야 함

### 3.2.8 통계적 검정력 Statistical Power
- 가설 검정에 대립가설이 참일 때 이를 올바르게 발견할 수 있는 으력
1. 통계적 검정력
- 대립가설이 참일 때 귀무가설을 기각할 확률
- 검정력이 높을수록 대립가설이 참일 때 이를 올바르게 검출할 가능성이 커짐
- 검정력 = 1 - β
2. 검정력에 영향을 미치는 요소
- 유의 수준 α
    - 유의 수준이 높아질수록 제1종 오류의 가능성이 높아지고, 이에 따라 검정력도 높아짐
    - 제1종 오류의 가능성을 증가시키는 트레이드오프를 의미
- 표본 크기 n
    - 표본 크기가 클수록 검정력은 높아짐
    - 더 큰 표본은 모집단의 특성을 더 정확하게 반영
    - 귀무가설을 기각할 가능성이 커짐
- 효과 크기 Effect Size
    - 효과 크기가 클수록 검정력도 커짐
    - 효과크기는 두 모집단 평균의 차이 or 변수 간 관계의 강도
    - 큰 효과 크기는 대립가설이 참일 가능성을 높이기 때문에 검정력도 증가
- 표준편차 σ
    - 모집단의 표준편차가 작을수록 검정력은 높아짐
    - 표준편차가가 작다는 것은 데이터의 변동성이 작다는 뜻
    - 이는 통계적으로 유의한 차이를 발견하기 쉬워짐
3. 검정력 분석
- 연구 설계 단계에서 필요한 표본 크기를 결정하거나, 연구 후 검정력의 충분성을 평가하는 데 사용
- 표본 크기 계산
    - 연구자가 설정한 유의 수준, 효과 크기, 검정력을 기준으로 적절한 표본 크기 결정할 수 있음
- 사후 검정력 분석
    - 연구가 완료된 후, 실제 효과 크기와 표본 크기를 바탕으로 검정력을 계산하여, 연구가 충분히 강력했는지 평가
4. 검정력 중요성
    - 통계적 검정력이 낮으면 대립가설이 참일 때 이를 올바르게 발견할 가능성이 낮아져 연구의 신뢰성이 떨어질 수 있음
