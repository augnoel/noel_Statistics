# 14. 통계적 검정력 분석

## 14.1 통계적 검정력의 개요
### 14.1.1 통계적 검정력statistical power의 정의
- 가설 검정에서 대립가설이 참일때 이를 올바르게 검정하여 귀무가설을 기각할 수 있는 확률
- 검정력은 제2종 오류β룰 범허자 얺울 확률
- 검정력 = 1 - β
- 가설 검정에서 얼마나 민감하게 실제 효과를 감지할 수 있는지 나타냄
- 검정력이 높다
    - 연구에서 실제 효과를 발견할 가능성이 높다는 것을 의미
    - 작은 효과 크기라도 감지할 수 있어 더 미세한 분석 가능
    - 연구에서 시간, 비용 등의 효율성 높일 수 있음
    - 연구의 신뢰성을 높여줌
### 14.1.2 통계적 검정력의 중요성
- 검정력이 낮다
    - 실제로 존재하는 효과를 놓치고 귀무가설을 기각하지 못할 가능성이 큼
    - 연구자는 제2종 오류를 범할 위험이 커짐
    - 유의미한 결과 얻기 어려움
### 14.1.3 검정력, 유의수준, 표본 크기 간의 관계
- 유의수준 α
    - 1종 오류를 범할 확률
    - 귀무가설이 참일 때 이를 잘못 기각할 확률
    - 일반적으로 0.05, 이 값을 낮추면 검정력이 감소할 수 있고 제 1종 오류의 가능성을 줄일 수 있지만, 그만큼 검정력도 감소
- 표본 크기 n
    - 검정력에 직접적인 영향을 미침
    - 표본 크기가 클수록 표본 평균의 분산이 줄어들어 실제 효과를 잘 탐지 할 수 있어 검정력 증가
- 효과 크기 effect size
    - 연구에서 실제로 존재하는 차이, 관계의 크기
    - 효과 크기가 클수록 표본 크기가 적더라도 쉽게 탐지할 수 있어 검정력 높아짐

## 14.2 검정력 계산
### 14.2.1 검정력 함수Power Function와 계산 방법
- 검정력 함수 Power Function
    - 가설 검정에서 대립가설이 참일 때, 귀무가설이 기각할 확률을 계산할 때 사용
    - 주로 통계적 검정에 사용되며, 특정 검정 통계량의 임계값을 넘는 확률

    

    - H_0 : 귀무가설, H_1 : 대립가설
- 검정력 계산 방법
1. 효과 크기 E
- 연구에서 기대되는 실제 효과 크기를 추정
-  Cohen's d 같은 지표를 사용하여 표현
2. 유의수준 α
- 일반적으로 0.05
- 제 1종 오류의 확률
3. 표본 크기 n
- 사용할 표본의 크기 설정
- 큰 표본 크기는 검정력을 높임
4. 표준 오차SE 계산
- 표본 크기와 표준 편차에 의해 결정
5. 검정 통계량 계산
- 주어진 조건하에서 검정 통계량 계산
6. 검정력 계산
- 검정 통계량을 바탕으로 검정력 계산
- 통계적 소프트웨어나 검정력 표를 이용해 계산
### 14.2.2 표본 크기와 효과 크기의 영향
- 표본 크기
    - 표본 크기가 증가하면 검정력도 증가
    - 그 이유는 표본 크기가 클수록 표본 평균의 표준 오차가 줄어들기 때문에 실제 효과를 감지하기 쉬워지기 때문
- 효과 크기
    - 효과 크기가 크면 검정력 증가
    - 큰 효과는 쉽게 탐지될 수 있기 때문에, 작은 표본 크기로 높은 검정력을 얻을 수 있음
### 14.2.3 소프트웨어를 이용한 검정력 계산
- G*Power
    - 모료로 사용 가능한 검정력 계산 소프트웨어
    - 다양한 통계 검정에서의 검정력을 계산 가능
    - 사용자는 효과 크기, 표본 크기, 유의수준을 입력하여 검정력 계산 가능
- R
    - 다양한 패키지를 통해 검정력 계산 지원
    - 'pwr' 패키지를 이용해 t-검정, X^2-검정, 회귀 분석 등 가능
- SPSS
    - 다양한 통계 분석 기능 제공
    - 사용자 인터페이스를 통해 쉽게 검정력 분석 수행
- SAS
    - 고급 통계 분석 소프트웨어
    - 검정력 분석을 포함한 다양한 분석 도구 제공
    - 프로그래밍 언어를 통해 사용자 지정 검정력 계산 가능

## 14.3 검정력 분석의 실제 적용
### 14.3.1 가설 검정에서의 검정력 분석
- 검정력은 대립가설이 참일 때 귀무가설을 기각할 수 있는 능력을 의미
- 검정력 분석이 가설 검정에 사용되는 이유
    - 연구의 타당성 확보
        - 적절한 검정력을 확보해야 하는데 낮은 검정력은 참인 대립가설을 검출 못할 위험을 증가시키므로 검정력 분석을 통해 필요한 표본 크기를 결정하여 연구의 신뢰성을 높임
    - 제2종 오류 β 오류 관리
        - 검정력은 제2종 오류의 보완 개념
        - 제2종 오류를 줄이기 위해 검정력을 높이는 것이 중요
        - 검정력이 높을수록 대립가설이 참일 때 귀무가설을 기각할 가능성이 커짐
- 연구 자원의 효율적 사용
    - 검정력 분석을 통해 필요한 표본 크기를 결정함으로써, 연구 자원을 효율적으로 사용 가능
    - 지나치게 큰 표본 크기는 자원의 낭비를 초래 가능
    - 반대로 표본 크기가 너무 작으면 검정력이 낮아져 연구 결과의 신뢰성이 떨어짐
### 14.3.2 연구 설계와 검정력 고려
- 표본 크기 결정
    - 크기는 검정력에 직접적인 영향을 미침
    - 검정력 분석을 통해 최적의 표본 크기 결정
    - 연구의 성공 확률을 높이는데 중요
- 유의수준α 선택
    - 검정력과 제1종 오류 간의 균형을 고려
    - 높은 유의수준은 검정력을 높이지만, 제1종 오류 위험을 증가시킴
- 효과 크기 추정
    - 실제로 기대되는 효과의 크기
    - 연구 설계 단계에서 효과 크기를 적절히 추정해야 검정력 분석이 의미있어짐
- 연구 목적과 맥락
    - 연구의 목적에 따라 검정력 요구사항이 달라질 수 있음
### 14.3.3 검정력 분석의 사례 연구
- 의학 연구에서 검정력 분석
    - 충분한 검정력 확보를 위해 연구 설계 단계에서 예상 효과 크기와 표본 크기를 기반으로 검정력 계산하여 임상시험의 성공 가능성을 높임
- 심리학 연구에서의 검정력 분석
    - 검정력 분석을 통해 필요한 최소 표본 크기를 결정함으로써 유의미한 결과를 얻을 가능성을 높임
- 교육 연구에서의 검정력 분석
    - 표본 크기와 효과 크기 최적화하여 연구가 의미있는 결론을 도출할 수 있도록 설계

## 14.4 검정력과 오류의 관계
### 14.4.1 제1종 오류α와 제2종 오류β
- 제1종 오류 α
    - 귀무가설이 참일 때 이를 잘못 기각하는 오류
    - 이 오류의 확률을 유의수준 α이라고 함
- 제2종 오류 β
    - 대립가설이 참일 때 귀무가설을 기각하지 못하는 오류
    - 연구의 검정력 1 - β과 밀접한 관련이 있음
### 14.4.2 오류의 상충 관계와 최적화
- 유의수준의 설정
    - α를 낮추면 제1종 오류의 위험을 줄일 수 잇지만 동시에 검정력도 감소하여 제2종 오류β가 증가할 수 있음
    - α를 너무 낮게 설정하면 연구의 검출력이 떨어져 대립가설을 검출하지 못할 가능성이 커짐
- 표본 크기 조정
    - 표본 크기를 늘리면 검정력이 증가하여 제2종 오류를 줄일 수 있음
    - 표본 크기를 무한정 늘릴 수는 없으므로, 연구 목적에 맞는 적절한 표본 크기를 결정하는 것이 중요
- 효과 크기 고려
    - 효과 크기가 크면, 같은 유의수준에서 검정력이 높아즈므로, 제2종 오류를 줄일 수 있음
    - 예상되는 효과 크기에 맞춰 연구 설계
### 14.4.3 오류율과 검정력의 균형
- 검정력 목표 설정
    - 일반적으로 연구에서 검정력을 80% 이상으로 설정
    - 제2종 오류를 20%이하로 유지하는 것을 목표
    - 연구가 의미있는 효과를 검출할 가능성을 높임
- 유의수준과 검정력의 트레이드오프
    - α와 β는 반비례 관계
    - 한쪽을 줄여 다른 쪽을 지나치게 희생하지 않도록 균형을 잡아야 함
- 표본 크기의 고려
    - 충분한 표본 크기는 검정력을 높여 제2종 오류를 줄이는 것이 중요
    - 효과 크기가 작을 때 큰 표본 크기를 통해 검정력을 확보해야 함
    - 표본 크기를 늘리면 연구 비용과 시간이 증가할 수 있으므로, 실질적인 한계 내에서 적절한 균형을 찾는 것이 중요
- 연구의 실질적 중요성
    - 통계적으로 유의한 결과가 항상 실질적으로 중요한 것은 아님
    - 검정력과 오류율의 균형을 맞출 때, 연구자가 추구하는 결과의 실질적 중요성을 고려하여 설정

## 14.5 고급 검정력 분석
### 14.5.1 다중 비교에서의 검정력
- 다중 비교 문제
    - 다중 비교 Multiple Comparisons
        - 여러 가설 검정을 동시에 수행할 때 발생하는 문제
    - 예시, 여러 그룹 간 차이를 비교하거나 여러 변수 간 관계 분석 때 각 검정에서 유의순을 개별적으로 설정하면 전체 오류율이 증가하는 문제가 생김
    - 다중 비교에 대한 가장 큰 위험은 제1종 오류 발생 확률이 높아진다는 것
- 다중 비교 보정
    - 다중 비교 문제를 해결하기 위해 보정 방법이 사용됨
       - 본페로니Bonferroni 보정
       - Tukey's HSD ; Honestly Significant Difference
       - Scheffé 방법
    - 전체 유의수준을 일정하게 유지하면서 개별 검정의 유의수준을 조정하여 제1종 오류 제어
- 검정력 영향
    - 다중 비교 보정 방법은 제1종 오류를 줄이는 데 효과적이지만, 보정을 통해 유의수준을 낮추면 검정력이 감소하는 경향이 있음
    - 중요한 효과를 놓칠 위험을 증가
    - 연구자는 다중 비교 상황에서 검정력을 충분히 확보하기 위해 적절한 표본 크기를 설정
    - 다중 비교 보정 강도를 조정하는 전략을 사용
### 14.5.2 베이지안 접근과 검정력 분석
- 베이지안 접근의 개요
    - 베이지안 통계학은 사전 확률과 사후 확률을 사용하여 데이터를 해석하는 방법
    - 주어진 데이터와 사전 지식을 결합하여 검정력을 분석
    - 전통적인 빈도주의 접근과 다르게 결과에 대한 불확실성을 더 명확하게 반영
- 베이지안 검정력
    - 사후 확률 분포에 기반하여 계산
    - 전통적인 검정력(빈도주의)에서 검정력은 특정 효과 크기를 가정하고 계산되지만, 베인지안 검정력은 다양한 효과 크기를 고려할 수 있어 보다 유연하고 현실적인 평가 가능
    - 베이지안 방법은 작은 표본 크기에서도 강력한 분석 도구 제공
- 베이지안 검정력 적용
    - 베이지안 검정력 분석은 특히 불확실성이 높은 상황이나 사전 정보가 중요한 역할을 하는 상황에서 유용
### 14.5.3 불확실성 하에서의 검정력 평가
- 불확실성과 검정력
    - 실제 연구에는 다양한 불확실성 요소
    - 표본의 변동성, 측정 오류, 모형의 적합성 등은 연구 겨로가의 검정력에 영향을 미침
    - 불확실성 하에서 검정력을 평가하는 것은 이러한 요소들을 고려하여 분석의 신뢰성을 높이는 과정
- 시나리오 분석
    - 불확실성 하에서 검정력을 평가하는 방법
    - 다양한 가정하에서 검정력을 평가하고 최악의 경우와 최선의 경우를 비교하여 연구 설계를 최적화하는 방법
- 로버스트 검정력 분석
    - 다양한 불확실성 하에서도 일정한 검정력을 유지할 수 잇는 방법을 연구
    - 표본 크기가 작거나 데이터가 왜곡된 경우 신뢰할 수 있는 결과 제공하는 방법 찾음
    - 특히 비정형적인 데이터나 비정규 분포 데이터에서 중요
