## 7.1 상관 분석의 기초

### 7.1.2 상관 계수의 종류
- 데이터의 특성과 분석 목적에 따라 적절한 상관 계수를 선택해야 함
1. 피어슨 상관 계수 Pearson Correlation Coefficient
- 가장 일반적으로 사용되는 상관 계수
- 두 변수 간의 선형 관계를 측정
- 두 변수는 연속형이어야 하고, 정규 분포를 따르는 것이 이상적
2. 스피어만 상관 계수 Spearman's Rank Correlation
- 순위를 기반으로 한 상관 계수
- 비선형 관계 또는 이상치Outlier가 있을 경우 유용
- 두 변수 간의 순서가 일관되게 증가하거나 감소하는 정도를 측정
3. 켄달의 타우 상관 계수 Kendall's Tau Correlation
- 순위를 기반으로 두 변수 간의 관계를 측정
- 작은 표본 또는 순서 데이터에 적합
- 순서 데이터 간의 일관성 또는 불일치성을 측정
### 7.1.3 상관 분석의 목적
- 두 변수 간의 관계를 이해하고, 이러한 관계가 통계적으로 유의미한지 판단하는 것이 목적
1. 변수 간의 관계 이해
- 두 변수 간의 상호 작용을 이해하고, 그 관계가 양의 관계인지 음의 관계인지 파악할 수 있음
2. 변수 선택
- 다중 회귀 분석Multiple Regression과 같은 모델링 기법에서 상관 분석을 통해 변수 간의 중복성을 확인하고, 모델에 포함할 변수 선택 가능
3. 예측 모델 개선
- 상관 분석을 통해 변수 간의 관계를 이용해 예측 모델의 성능을 개선할 수 있음
- 강한 상관 관계를 가지 변수는 예측력 있는 변수로 활용될 가능성이 큼
4. 연구 가설 검정
- 연구에서 제시된 가설이 두 변수 간의 관계에 대한 것이라면, 상관 분석을 통해 가설 검정 가능 

## 7.2 상관 계수의 계산
- 상관 계수는 두 변수 간의 관계를 정량적을 측정하는 지표
### 7.2.1 피어슨 상관 계수 Pearson Correlation Coefficient
- 두 연속형 변수 간의 선형 관계를 측정하는 방법
- 두 변수 간의 공분산을 각 변수의 표준 편차로 나눈 값



### 7.2.2 스피어만 상관 계수 Spearman's Rank Correlation
- 순위 데이터를 기반으로 두 변수 간의 단조 관계(순서가 일관되게 증가, 감소하는 관계)를 측정
- 비선형 관계에도 사용 가능
- 이상치의 영향 덜 받음
- 순위로 변환된 데이터에 피어슨 상관 게수를 적용해 계산



### 7.2.3 켄달의 타우 상관 계수 Kendall's Tau Correlation
- 두 변수 간의 일관된 순선 관계를 측정
- 순위 기반의 상관 계수
- 데이터의 일관성을 평가
- 작은 표본 크기 또는 순서형 데이터에 적합
- 스피어만 상관 계수보다 좀 더 강건하게 일관성 측정
- 순위 쌍의 일관성(순서가 동일하거나 불일치한 쌍)을 계산하여 표현



## 7.3 상관 계수의 해석
### 7.3.1 상관 계수의 크기와 방향
- 크기
    - +1에 가까운 값은 두 변수 간 강한 양의 선형 관계
    - -1에 가까운 값은 강한 음의 선형 관계
    - 0에 가까운 값은 선형 관계가 거의 없거나 전혀 없음
- 방향
    - 양수 : 한 변수가 증가할 때 다른 변수도 증가하는 경향
    - 음수 : 한 변수가 증가할 때 다른 변수는 감소하는 경향
### 7.3.2 상관과 인과관계의 구분
- 상관 관계
    - 상관 계수는 두 변수 간의 관계의 뱡향과 크기를 나타내지만, 인과적causal인지 알 수 없음
- 인과 관계
    - 한 변수의 변화가 다른 변수에 직접적인 영향을 미친다는 것을 의미
    - 상관 관계만으로 인과 관계 증명 불가
    - 인간 관계를 입증하려면 추가적인 분석이나 실험적 연구 필요
- 주의사항
    - 상관 계수가 높거나 낮다고 해서 반드시 두 변수 간에 인관 관계가 존재하는 것 아님
### 7.3.3 상관 계수의 한계와 주의사항
- 한계
    - 선형 관계만 측정
        - 피어슨 상관 계수는 선형 관계만 측정하므로 비선형 관계를 놓칠 수 있음
        - 비선형 관계에서는 낮은 상관 계수가 나올 수 있지만 여전히 변수 간의 강한 관계가 존재할 수 있음
    - 이상치
        - 상관 계수는 이상치에 민감
    - 상관의 통계적 유의성
        - 상관 계수가 통계적으로 유의미하지 여부는 표본 크기에 따라 달라짐
        - 큰 표본에서는 작은 상관 계수도 유의미할 수 있지만, 작은 표본에서는 큰 상관 계수 조차 유의미하지 않을 수 있음
    - 변수의 스케일
        - 상관 계수는 변수의 스케일에 영향을 받지 않음
        - 대신데이터의 변동성을 완전히 설명 못 할 수도 있음
- 주의사항
    - 상관 계수를 해석할 때 데이터의 특성과 연구 맥락을 충분히 고려할 것
    - 상관 분석 후에는 항상 데이터 시각화를 통해 상관 계수가 나타내는 관계를 시각적으로 확인해 볼 것
    - 상관 계수 외에도 비선형 관계를 탐색할 수 있는 다른 통계적 방법 고려해 볼 것
  
## 7.4 상관 분석의 실제 적용
- 데이터 간 관계를 이해하고, 모델링, 데이터 분석의 여러 단계에서 활용하는 도구
### 7.4.1 상관 행렬 Correlation Matrix
- 여러 변수들 간의 상관 계수를 행렬 형태로 나타낸 것
- 각 행과 열은 변수에 대응
- 각 셀에는 해당 변수 쌍의 상관 계수가 포함
- 데이터 셋의 모든 변수 간의 상관 관계를 한눈에 파악할 수 있음
### 7.4.2 상관 분석을 통한 변수 선택
- 데이터 분석, 모델링을 할 때, 모든 변수를 사용할 필요 없음
- 상관 분석을 통해 변수 간의 관계를 이해하고, 분석에 중요한 변수 선택
- 상관 관계가 높은 변수들 중 일부는 중복된 정보를 제공할 수 있음
- 모델의 복잡성을 줄이기 위해 상관 분석을 통해 변수 선택 가능
- 방법
    1. 상관 행렬 분석
       - 높은 상관 계수를 가진 변수 쌍을 식별
       - 그 중 하나를 제거하거나 다른 방식으로 결합 가능
    2. 다중공선성Multicollinearity 제거
       - 상관 계수가 높은 변수는 다중공선성을 초래할 수 있음
       - 이를 피하기 위해 일부 변수를 제외할 수 있음
### 7.4.3 상관 계수의 시각화 (산점도, 히트맵 등)
- 상관 계수를 시각화하면 데이터 간의 관계를 더 명확하게 이해할 수 있음
- 시각화를 통해 데이터 패턴을 쉽게 파악하고 분석 결과를 직관적으로 전달 가능
1. 산점도 Scatter Plot
   - 두 변수 간 관계를 점으로 표시
2. 히트맵 Heatmap
   - 상관 행렬을 색상으로 표현한 시각화 도구
   - 상관 계수의 크기에 따라 색상이 달라져, 변수 간의 상관 관계를 시각적으로 쉽게 비교 가능
3. 페어플롯 Pairplot
   - 여러 변수들 간의 산점도를 한 번에 볼 수 있는 시각화 도구
   - 변수 간의 관계를 종합적으로 이해하는 데 유용

## 7.5상관 분석의 고급 주제
### 7.5.1 부분 상관 Partial Correlation
- 두 변수 간의 상관 관계를 다른 변수의 영향을 제거한 후에 계산하는 방법
- 일반적인 상관 분석에서는 모든 변수가 포함되지만, 부분 상관 분석에서는 특정 변수를 통제하여 두 변수 간의 순수한 관계를 평가
- 예시,  변수 X와 Y 사이의 상관 관계를 분석하고자 할 때, 다른 변수 Z의 영향을 제거하고 싶다면 부분 상관을 사용. 이는 
X와 Y 간의 상관이 실제로 Z에 의해 매개된 것인지, 아니면 X와 Y 자체 간의 관계인지를 구분하는 데 유용



### 7.5.2 다변량 상관 분석
- 여러 변수 간의 복잡한 상관 관계를 동시에 분석하는 방법
- 여러 변수들이 서로 어떻게 관계를 맺고 있는지 보다 심층적으로 이해하는데 도와줌
- 캐논 상관 분석 Canonical Correlation Analysis ; CCA
    - 두 개의 변수 그룹 사이의 선형 관계를 측정
    - 이 변수들이 서로 어떤 식으로 관계를 맺고 있는지 파악 가능
    - 각 변수 세트에서 주성분을 추출하여 이들 간의 상관 계수 계산
### 7.5.3 상관 계수의 통계적 검정
- 상관 계수는 표본 데이터를 기반으로 계산
- 이 상관이 우연에 의한 것인지 아니면 실제로 의미 있는지 검정이 필요
- 상관 계수의 통계적 검정은 상관 계수가 유의미하지 판다
- 가설 설정
    - 귀무가설 H_0 : 상관 계수 ρ=0, 변수 간의 상관 관계 없음
    - 대립가설 H_1 : 상관 계수 ρ≠0, 변수 간 상관 관계 있음
- t-검정 이용
    - 피어슨 상관 계수r에 대해 t-검정을 사용하여 검정 통계량 계산

     

    - 이 검정 통계량을 통해 t-분포에서 p-값을 얻고, 상관 계수의 유의성 판단
