# 15. 다중 비교 문제 Multiple Comparisons Problem

## 15.1 다중 비교 문제의 개요
### 15.1.1 다중 비교 문제의 정의
- 한 연구에서 여러 개의 통계적 가설을 동시에 검정할 때 발생하는 문제
- 여러 그룹 간 평균 차이를 비교
- 다수의 독립 변수가 종속 변수에 미치는 영향 분석
- 각각의 검정에서 유의수준을 개별적으로 설정하면 전체적인 오류율이 증가
    - 각 검정에서 발생할 수 있는 제1종 오류 α가 누적도어 연구 전체의 신뢰성 저하
- 여러 검정의 동시 수행
    - 한 연구 내에서 여러 가설을 동시에 검정함으로써 각 검정의 독립성이 보장되지 않을 수 있음
- 오류율의 증가
    - 개별 검정의 제1종 오류 확률을 그대로 유지할 경우, 전체 연구에서의 제1종 오류 발생 확률이 증가
- 결과 왜곡
    - 다수의 검정 결과 중 일부가 우연에 의해 유의미하게 나타날 가능성이 높아져, 실제로는 존재하지 않는 효과를 발견하게 됨
### 15.1.2 다중 비교로 인한 오류와 왜곡
1. 제1종 오류의 중가
- 귀무가설이 참일 때 이를 잘못 기각하는 오류
- 다중 비교의 영향
    - 여러 검정을 동시에 수행하면, 각 검정마다 제1종 오류가 발생할 확률이 존재
- 연구 결과 중 일부가 실제로는 무의미한 효과임에도 불구하고 유의미하게 나타날 수 있어, 잘못된 과학적 결로능ㄹ 도출할 위험이 있음
2. 제 2종 오류의 영향
- 대립가설이 참일 때 귀무가서을 기각하지 못하는 오류
- 다중 비교의 영향
    - 다중 비교 보정을 통해 제1종 오류를 통제하면, 유의수준이 더욱 엄격해져 제2종 오류의 확률이 증가할 수 있음
    - 실제로 존재하는 효과를 놓칠 가능성이 높음
- 연구의 민감도 저하
    - 효과를 탐지할 수 있는 능력이 감소
3. 다중 비교로 인한 왜곡
- 유의미한 결과의 남용
    - 다수의 검정 중 일부가 우연에 의해 유의미한 결과를 보이면서 전체 연구 결과가 왜곡될 수 있음
- 해석의 복잡성 증가
    - 다중 비교로 인한 오류의 증가로 인해 연구 결과를 해석하고 일반화하는 데 어려움이 따를 수 있음
### 15.1.3 다중 비교 문제의 중요성
1. 연구의 신뢰성 확보
- 정확한 결론 도출
    - 다중 비교 문제를 적절히 관리하지 않으면, 잘못된 결론을 도출할 가능성이 커짐
    - 연구의 신뢰성 저하
- 재현 가능성
    - 잘못된 긍정 결과는 다른 연구에서 재현될 가능성을 낮춰 과학적 지식의 축적에 장애
2. 통계적 타당성 유지
- 오류 통제
    - 다중 비교 보정 방법을 통해 제1종 오류를 통제함으로써, 통계적 타다성을 유지 가능
- 연구 설계의 질 향상
    - 다중 비교 문제를 인식하고 이를 해결하기 위한 적절한 방법을 적용함으로써, 연구 설계의 질을 향상시킬 수 있음
3. 과학적 윤리 준수
- 잘못된 정보의 방지
    - 다중 비교 문제를 간관하면, 잘못된 정보를 바탕으로 한 정책 결정이나 실무적 적용이 이뤄질 수 있음
- 투명한 연구 수행
    - 오류를 관리하고 연구 결과의 신뢰성을 높이는 것을 과학적 윤리의 중요한 부분
4. 연구 자원의 효율적 사용
- 불필요한 추가 연구 방지
    - 다중 비교 문제로 인해 잘못된 결론이 도출되면 이를 수정하기 위한 추가 연구가 필요
    - 이는 시간과 비용의 낭비로 이어질 수 있음
- 효과적인 자원 배분
    - 정확한 연구 결과를 통해 자원을 효과적으로 배분하고, 중요한여구 주제에 집중할 수 있음

## 15.2 다중 비교 문제 해결 방법
### 15.2.1 보수적인 유의수준 설정
- 다중 비교로 인한 오류를 줄이기 위해 개별 검정에서 사용하는 유의수준을 낮추는 방법
- 일반적으로 사용되는 유의수준은 0.05이지만, 다중 비교 문제를 해결하기 위해 0.01, 0.001로 설정 가능
- 유의수준을 낮추면, 개별 감정에서 귀무가설을 기각할 기준이 더 엄격해지므로, 제1종 오류 가능성 낮아짐
- 유의수준을 너무 낮추면, 제2종 오류(실제로 존재하지만, 이를 발견하지 못할 오류)의 가능성이 증가할 수 있음
- 보수적인 유의수준 설정은 제1종 오류를 줄이는 반면, 통계적 검정력(효과를 탐지할 수 있는 능력)을 감소시킬 수 있음
### 15.2.2 다중 비교 보정 기법
1. 본페로지 보정 Bonferroni Correction
- 가장 간단한 보정 방법
- 전체 유의수준을 개별 검정의 수로 나누어 각 검정의 유의수준으로 설정하는 방식
- 예시, 5개릐 검정을 동시에 수행할 때, 전체 유의수준을 0.05로 설정하고 이를 5로 나눠 각 검정의 유의수준을 0.01로 설정
2. 튜키의 HSD ; Tukey's Honest Significant Difference
- 모든 그룹 간 비교를 동시에 고려하는 방법
- ANOVA 이후에 사용
- 각 그룹간 평균 차이가 어느 정도로 커야 유의미한지 결정
- 다중 비교를 위한 매우 효과적인 방법
- 여러 그룹 간 차이를 동시에 고려하여 유의수준 보정
- ANOVA 후에 그룹 간 차이를 분석할 때 유용
3. 셰페 Scheffé 방법
- 가장 보수적인 다중 비교 방법
- 특히 모든 가능한 그룹 간 비교를 고려
- 다른 보정 밥버에 비해 유의수준이 매우 엄격하게 적용
- 유연하게 적용할 수 있으며, 다중 비교 상황에서의 신뢰성 높임
- 너무 보수적이어서, 신제로 의미있는 차이를 탐지하는 능력이 떨어질 수 있음
4. 홀름-본페로니 방법 Holm-Bonferroni Method
- 본페로니 보정을 개선한 방법
- 순차적으로 검정의 유의수준을 조정
- 가장 작은 p-값부터 시작하여 각각의 p-값에 대해 유의수준 점진적으로 늘리는 방식
- 본페로니 보정보다 덜 보수적
- 제2종 오류를 줄이는 데 도움
### 15.2.3 사전 계획된 비교와 사후 비교의 구분
1. 사전 계획된 비교 Planned Comparisons
- 연구가 시작되기 전에 특정 비교를 미리 계획하고 해당 비교에 대해 검정할 것으로 명확히 설정하는 방식
- 특정 두 그룹 간 차이를 비교하기 위해 미리 계획된 검정을 수행하는 것이 해당
- 다중 비교 문제를 줄일 수 있으며, 연구 결과의 타당성을 높임
- 계획된 비교는 연구의 주요 가설을 검증하는 데 초점을 맞추므로, 보정이 덜 필요할 수 있음
- 연구의 주된 가설을 검정하는 데 유리하며, 불필요한 비교를 최소화할 수 있음
2. 사후 비교 Post Hoc Comparisons
- 연구 수행 후에 예기치 않은 결과에 대해 추가적인 비교를 수행하는 방식
- 연구자가 특정 그룹 간 차이가 유의미하게 나타났을 때, 그 원인을 규명하기 위해 추가적으로 실시하는 비교
- 예기치 않은 결과를 탐구하거나, 새로운 가설을 설정하기 위해 필요
- 다중 비교 문제를 더욱 복잡하게 만들 수 있으므로, 보정 기법을 반드시 적용해야 함
- 사후 비교에서는 특히 본페로니 보정이나 튜키의 HSD 방법과 같은 다중 비교 보정 기법 필수

## 15.3 다중 비교 보정 기법
### 15.3.1 본페로니Bonferroni 보정
- 전체 실험에서 허용할 수 있는 제1종 오류의 총합이 미리 정해진 유의수준을 초과하지 않도록, 개별 검정에서 사용하는 유의수준을 조정하는 방법
- 개별 유의수준은 전체 유의수준을 검정의 개수로 나눠 계산



- 적용이 간단하며, 보수적으로 제1종 오류를 통제 가능
- 검정의 개수를 많아지면, 각 검정에서 유의성을 얻기 어려워져 제2종 오류가 증가할 수 있음
### 15.3.2 Tukey's HSD ; Honest Significant Difference
- ANOVA 후에 모든 가능한 그룹 간 평균 차이를 비교할 때 사용
- 특히 동일한 표본 크기를 가진 그룹 간 비교에 적합
- 두 그룹 간 평균 차이가 유의미하게 다른지를 판단하기 위한 임계값 계산
- 그룹 간 모든 가능한 쌍을 고려하여 계산
- HSD값은 두 평균 간 차이를 검정할 때 사용되는 분포
- 샘플 수와 그룹 수에 따라 결정
- 보수적이면서도 효율적인 방벙
- 다수 그룹 간 비교 시 제1종 오류를 효과적으로 제어 가능
- 특히 ANOVA 후에 여러 그룹 간 평균 비교할 때 사용
### 15.3.3 Scheffé 방법
- 모든 가능한 그룹 간 비교에 대해 유의수준 통제
- 그룹 간 비교뿐만 아니라, 선형 결합과 같은 복잡한 비교에도 사용될 수 있음
- f-분포를 이용하여, 모든 가능한 비교에 대해 유의수준을 보수적으로 조정
- 가장 다양한 비교를 허용하지만, 동시에 가장 보수적인 접근법
- 매우 보수적인 방법으로 다중 비교로 인한 제1종 오류를 철저히 방지
- 모든 가능한 비교에 대해 유의미성 보장
- 지나치게 보수적이기 때문에, 실제로 유의미한 효과를 탐지하는 능력이 떨어질 수 있음
- 복잡한 비교나, 유연성이 필요한 연구에서 사용됨 
### 15.3.4 Holm-Bonferroni 방법
- 단계별 유의 수준을 조정해 나가면서 보다 유연하게 오류를 통제
- p-값이 작은 순서대로 검정을 진행하며, 각 검정에서 본페로니 방식으로 유의수준 조정
- 그러나 첫번째 검정에서 가장 작은 p-값에 대해 전체 유의수준을 사용하고, 이후 단계마다 조금씩 줄여나감
- 절차
    1. 각 가설에 대한 p-값 정렬
    2. 첫번쩨 가설의 p-값은 α/m(m은 가설 수)와 비교
    3. 두번째 가설의 p-값은 α/(m-1)과 비교
    4. 이 과정을 계속 진행하며, 유의수준이 충족될 때까지 비교를 멈춤
- 본페로니 보정보다 덜 보수적이며, 재1종 오류와 제2종 오류 간 균형을 보다 잘 맞출 수 있음
- 본페로니 보정보다 효율적이면서도 신뢰성 유지하고 할 때 사용
### 15.3.5 FDR False Discovery Rate 조정
- 다중 비교 시 여러 가설 중 제1종 오류의 비율을 제어하는 방법
- 특히 연구에서 발견된 것들이 모두 진실일 가능성을 고려
- 보정하지 않은 방식보다 덜 보수적인 접근법 제공
- FDR은 발견된 통계적 유의성을 가진 결과 중에서 실제로 제1종 오류일 확률을 제어하는 방법
- 특히 가설을 동시에 검정하는 경우, 보수적인 방법들이 지나치게 엄격할 때 유용
- Benjamini-Hochberg 절차
    1. 모든 p-값을 오름차순으로 정렬
    2. 각 p-값에 대해 i/m*α(i는 p-값의 순서, m은 전체 검정 수)
    3. 가장 큰 유의미한 p-값을 기준으로 해당 검정과 그보다 작은 모든 검정이 유의미한 것으로 간주
- 다수의 가설을 검정할 때, 특히 탐색적 연구에서 유용하며, 발견율을 높이면서도 오류를 통제할 수 있음
- 제1종 오류율이 다소 높아질 수 있으며, 모든 연구 설계에 적합하지 않을 수 있음

## 15.4 다중 비교 문제의 실제 적용
- 다중 비교를 적절히 처리하지 않으면 연구 결과의 신뢰성이 크게 떨어질 수 있음
### 15.4.1 연구 디자인에서의 다중 비교 고려
- 사전 계획
    - 연구 디자인 단계에서 예상되는 다중 비교의 개수를 파악하고, 적절한 보정 방법 계획하는 것이 중요
    - 실험을 설계할 때 비교할 그룹의 수, 변수의 수 등을 미리 결정하고, 이에 딸ㄴ 다중 비교 문제 예상 가능
- 집단 간 비교 계획
    - 집단 간 비교를 미리 계획함으로써 사전 검정과 사후 검정 구분 가능
    - 사전 계획된 비교는 보정이 덜 필요할 수 있지만, 사후 검정은 더 강력한 보정이 요구
- 유의수준 설정
    - 다중 비교를 고려해 연구 전반에 걸친 유의수준을 설정하고, 개별 비교에서의 유의수준을 조정하는 전략 수립
    - 본페로니 보정과 같은 간단한 방법부터 정교한 보정 방법까지 선택 가능
- 연구 목적과 방법에 따른 선택
    - 연구의 목적이 탐색적인지, 확인적인지에 따라 다중 비교를 처리하는 방식이 달라질 수 있음
    - 탐색적 연구에서는 발견 가능성을 높이는 FDR 방법이 유리할 수 있으며, 확인적 연구에서는 보수적인 보정 방법이 더 적합할 수 있음
### 15.4.2 소프트웨어를 활용한 다중 비교 보정
- 주요 통계 소프트웨어
    - R, Python, SPSS, SAS, Stata 등 다양한 소프트웨어가 다중 비교 보정 기능을 제공
    - 각 소프트웨어에는 Tukey's HSD, Bonferroni 보정, Holm-Bonferroni 방법, FDR 조정 등 다양한 보정 기법이 내장
### 15.4.3 다중 비교 문제의 사례 연구
- 의학 연구 
- 심리학 연구
- 유전자 연구
- 경제학 연구

## 15.5 다중 비교와 통계적 검정력
### 15.5.1 다중 비교와 검정력의 관계
- 검정력 statistical power
    - 실제로 효과가 있는 경우 이를 올바르게 탐지할 확률
    - 다중 비교를 수행할 때, 여러 테스트를 동시에 진행하면서 제1종 오류를 통제하기 위해 유의수준을 낮추는 경우가 많음
    - 유의수준을 낮추면 검정력이 감소할 수 있움
- 유의수준 조정
    - 다중 비교에서 제1종 오류를 줄이기 위해 보수적인 유의수준을 설정하면, 검정력이 감소
    - 실제로 존재하는 효과를 발견하지 못할 가능성(제2종 오류)이 높아짐
- 검정력 감소의 이유
    - 검정력은 효과 크기, 유의수준, 표본 크기에 따라 결정
    - 다중 비교에서 유의수준을 낮추면 동일한 효과 크기와 표본 크기에서 검정력이 낮아짐
    - 다중 비교를 많이 수행할수록 개별 테스트의 검정력이 낮아져서 실제로 중요한 차이나 효과를 놓칠 가능성이 커짐
- 다중 비교 문제와 대안
    - 다중 비교로 인한 검정력 저하를 방지하기 위해 다양한 대안이 사용될 수 있음
    - 효과 크기를 늘리거나, 표본 크기를 증가시키는 방법으로 검정력을 높일 수 있음
### 15.5.2 다중 비교 보정 후 검정력 평가
- 보정 방법에 따른 검정력 변화
    - Tukey's HSD, Bonferroni 보정, FDR 조정 등 여러 보정 방법이 있지만, 이들 중 읿는 검정력을 크게 낮추기도 함
- 검정력 평가
    - 보정 후 검정력을 평가하기 위해 다양한 방법이 사용될 수 있음
    - 소프트웨어를 사용하여 각 테스트 검정력을 계산하거나, 시뮬레이션을 통해 보정 전후의 검정력 비교 가능
- 소프트웨어를 활용한 검정력 분석
    - 통계 소프트웨어는 보정 후 검정력을 자동으로 계산할 수 있는 기능 제공
    - 연구자가 보정 전후의 검정력을 정확히 평가하고, 연구 설계나 분석 방법 조정
### 15.5.3 다중 비교와 통계적 유의성의 균형
- 다중 비교 문제를 다루면서 검정력과 통계적 유의성 간의 균형이 중요
- 통계적 유의성과 검정력 간의 균형
    - 다중 비교 보정 후에도 통계적 유의성과 검정력 간의 균형을 유지해야 함
    - 보정 방법이 지나치게 엄격하면 실제로 존재하는 효과를 발견하기 어려워지며, 지나치게 느슨하면 잘못된 결론에 도달할 수 있음
- 사전 계획과 검정력 고려
    - 연구를 설계할 때, 다중 비교로 인한 검정력 저하를 최소화할 수 있는 계획을 세우는 것이 중요
    - 사전 계획된 비교를 우선시하고, 사후 검정을 최소화하는 방법이 있음
    - 적절한 보정 방법을 선택하여 유의수준과 검정력 간의 균형 필요
- 효과 크기와 표본 크기 조정
    - 효과 크기가 충분히 크거나, 표본 크기가 충분히 크다면, 보정 후에도 검정력을 유지할 수 있음
    - 연구 설계 단계에서 효과 크기와 표본 크기를 고려해 다중 비교로 인한 검정력 저하 방지

## 15.6 다중 비교의 한계와 대안
### 15.6.1 다중 비교 문제의 한계
- 다중 비교 문제는 여러 통계 검정을 동시에 수행할 때 제1종 오류 확률이 증가하는 것을 의미
- 제1종 오류율의 증가
    - 다중 비교를 수행할 때 개별 검정의 유의수준을 유지한다고 하더라도, 전체 실험에서의 제1종 오류율이 크게 증가
    - 실험에서 유의하지 않은 차이나 효과를 유의하다고 잘못 판단할 위험을 높임
- 검정력 감소
    - 다중 비교 문제를 해결하기 위해 보수적인 유의수준 조정 방법을 사용할 경유, 검정력이 감소하여 실제로 존재하는 효과를 발견하기 어려워질 수 있음
- 실험이 복잡성 증가
    - 여러 가설을 동시에 검정하다 보면, 실험 설계와 데이터 해석이 복잡해지며, 결과 해석이 어렵게 됨
    - 연구자가 잘못된 결론에 도달할 위험
- 해석의 어려움
    - 다중 비교 후 도출된 유의성은 단위 비교와 다른 위를 가질 수 있음
    - 보정 방법이 적용된 결과는 직관적으로 해석 어려움
### 15.6.2 다중 비교 회피 전략
- 사전 계획된 가설 검정
    - 연구 설계 단계에서 비교하고자 하는 가설을 사전에 명확히 계획
    - 계획에 따라 가설 검정을 수행하는 것이 중요
    - 사후 다중 비교의 필요성 줄일 수 있음
- 조정된 유의수준 설정
    - 다중 비교를 수행할 때, Bonferroni 보정 등 보수적인 유의수준 조정 방법을 사용하여 제1종 오류를 통제 가능
    - 검정력 저하를 방지하기 위해 필요한 경우 보정의 강도 조정 가능
- 효과 크기와 표본 크기 조정
    - 다중 비교로 인한 검정력 감소를 방지하기 위해 효과 크기를 증가시키거나, 표본 크기를 늘리는 것이 효과적일 수 있음
    - 충분한 표본 크기를 확보하면 다중 비교로 인한 오류 줄일 수 있음
- 차원 축소
    - 데이터의 차원을 줄이는 방법도 다중 비교 문제 회피 전략
    - 주성분 분석이나 요인 분석 같은 기법을 사용하면 비교해야할 변수의 수를 줄여 다중 비교의 필요성 감소시킬 수 있음
### 15.6.3 Bayesian 방법을 통한 대안적 접근
- 사전 확률의 도입
    - Bayesian 방법에서는 사전 확률을 도입하여 데이터 이전의 정보를 반영
    - 다중 비교 상황에서 개별 검정의 신뢰성을 높일 수 있음
    - 사전 확률을 통해 개별 가설이 참일 가능성을 고려함으로써, 다중 비교롸 인한 제1종 오류 줄일 수 있음
- 포스트리어 확률 계산
    - 데이터 관찰 후, 사후 확률을 계산하여 각 가설의 타당성 평가
    _ 다중 비교 문제를 겪는 빈도주의 접근보다 더 유연하고 직관적인 해석 가능
- 베이지안 모델 선택
    - Bayesian 접근에서는 모델 간 비교를 통해 가장 타당한 모델 선택하는 과정 포함
    - 다중 비교의 필요성이 자연스럽게 중러들며, 다중 비교 문제를 해결 가능
- MCMC 방법과 Bayesian 추정
    - 복잡한 모형에서 Bayesian 추정을 위해 MCMC(Markov Chain Monte Carlo) 방법이 사용
    - 고차원의 다중 비교 문제를 효율적으로 해결 가능
