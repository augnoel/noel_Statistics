# 06 신뢰구간 Confidence Intervals

## 6.1 신뢰구간의 기초
### 6.1.1 신뢰구간의 정의
- 모집단의 모수(평균, 비율 등)를 추정하는 데 사용되는 범위
- 표본 만을 이용해 모집단의 결과를 추정할 때, 모집단 평균등이 특정 범위 내에 있을 것이라 추정하는 범위
- 신뢰구간은 특정 신뢰순준에서 계산
    - 95% 신뢰수준에서 신뢰구간을 계산하면, 전체 모집단에서 추출한 표본 중 95%가 이 구간에 모수가 포함될 것이라는 의미
- 신뢰구간 자체가 변동
    - 표본을 여러 번 추출하면 각 표본에서 계산한 신뢰구간이 달라질 수 있음
    - 그 중 95%는 실제 모수를 포함
### 6.1.2 신뢰수준 Confidence Level
- 신뢰구간을 계산할 때 설정하는 확률
- 해당 확률에 따라 모수가 신뢰구간 내에 포함될 가능성을 나타냄
- 일반적으로 신뢰수준은 90%, 95%, 99%가 사용됨
- 높은 신뢰수준
    - 신뢰구간이 넓어짐
    - 모집단의 참값이 이 구간 내에 있을 가능성이 더 높음
- 낮은 신뢰수준 
    - 신뢰구간이 좁아짐
- 신뢰수준 : 1 - α (α : 유의수준)

## 6.2 신뢰구간 계산 방법
### 6.2.1 평균에 대한 신뢰구간
- 모집단 평균을 추정하기 위해 표본 평균을 사용하여 계산하는 구간
- 신뢰구간은 모집단 평균이 특정 구간 내에 있을 확률을 제공





### 6.2.2 비율에 대한 신뢰구간
- 특정 사건이 발생할 비율을 추정하기 위해 표본 비율을 사용하여 계산하는 구간



### 6.2.3 차이에 대한 신뢰구간
- 두 그룹 간의 차이를 추정하기 위해 사용



### 6.2.4 분산에 대한 신뢰구간
- 모집단의 분산을 추정하ㅣ 위해 표본 분산을 사용하여 계산하는 구간



## 6.3 신뢰구간과 표본 크기
### 6.3.1 표본 크기에 따른 신뢰구간 변화
- 신뢰구간은 표본 크기에 따라 변함
- 일반적으로 표본 크기가 커질수록 신뢰구간이 좁아지고, 작을수록 신뢰구간이 넗어짐
- 표본 크기가 커질수록 모집단의 평균을 보다 정확하게 추정할 수 있다는 것



- 표본 크기가 작을수록 신뢰구간이 넓어져, 추정의 불확실성이 커짐
- 표본 크기가 클수록, 신뢰구간 좁아져, 모집단 평균에 대한 추정이 더 정확해 짐
### 6.3.2 신뢰구간의 해석과 실질적 의미
- 정확한 의미
    - 신뢰구간은 특정 신뢰수준에서 모집단의 모수가 신뢰구간 내에 포함될 것이라는 통계적 확률을 제공
- 실질적 의미
    - 신뢰구간은 데이터를 기반으로 한 의사결정에 중요한 역할
    - 좁은 신뢰구간은 더 신뢰할 수 있는 추정
    - 넓은 신뢰구간은 불확실설이 크다는 것을 의미

## 6.4 신뢰구간의 실제 적용
- 신뢰구간은 모집단의 모수를 추정하는 도구
- 통계적 추론과 가설 검정에서 중요한 역할
### 6.4.1 신뢰구간을 이용한 추론
- 모수 추정
    - 신뢰구간을 사용하면 모집단의 모수 추정 가능
    - 표본 평균을 사용해 모집단 평균의 신뢰구간을 구하면, 모집단 평균이 일정 범위 내에 있을 가능성을 평가
- 의사결정
    - 신뢰구간을 통해 데이터 기반의 의사결정 가능
    - 두 제품의 평균 성능을 비교하는 연구에서 두 제품의 평균 차이에 대한 신뢰구간이 0을 포함하지 않는다면, 두 제품의 성능이 통계적으로 유의미하게 다루다는 결론
### 6.4.2 신뢰구간과 가설 검정
- 가설 검정과의 연계   
    - 신뢰구간은 가설 검정과 밀접하게 연관되어 있음
    - 귀무가설을 기각할 여부를 판단할 때, 해당 가설이 신뢰구간에 포함되는지 확인 가능
- 양측 검정과 신뢰구간
    - 예를 들어, 신뢰구간이 0을 포함하지 않는다면, 이는 5%유의수준에서 귀무가설 기각 가능
### 6.4.3 신뢰구간의 한계
- 모형 가정에 의존
    - 신뢰구간은 모형 가정에 크게 의존
    - 특히 정규성 가정이 만족되지 않으면, 신뢰구간이 왜곡될 수 있음
- 오해의 소지
    - 신뢰구간은 종종 모집단의 모수가 신뢰구간 안에 있을 확률로 오해될 수 있음
    - 신뢰구간은 동일한 조건 하에서 반복된 표본 추출 시, 그 중 일정 비율의 신뢰구간이 참 모수를 포함할 것이라는 의미
- 신뢰수준의 선택
    - 신뢰수준은 임의적으로 선택된 것
    - 너무 높은 신뢰수준은 신뢰구간이 넓어져 실질적인 유용성이 감소할 수 있음
    - 너무 낮은 신뢰수준은 신뢰구간이 좁아지면서 참 모수를 포함하지 못할 가능성도 있음
