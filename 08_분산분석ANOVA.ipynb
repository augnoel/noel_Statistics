# 08 분산분석 ANOVA

## 8.1 분산분석의 기초
### 8.1.1 분산분석의 정의
- 여러 그룹 간 평균 차이를 비교하기 위해 사용되는 통계적 방법
- 두 그룹 간 단순 평균 차이 비교하는 t-검정과 달리, 분산분석은 3개 이상의 그룹 간 평균을 비교할 때 유용
- 각 그룹 '내'의 데이터 분산과 그룹 '간'의 데이터 분산을 비교하여, 그룹 간 평균이 유의미하게 다른지를 평가
- 만약 그룹 간 분산 > 그룹 내 분산, 이는 그룹 간 평균이 다르다고 결론지을 수 있음
### 8.1.2 분산분석의 목적과 활용
- 목적
    - 여러 그룹 간 평균이 유의미하게 다른지 평가
    - 실험 설계에서 처리 효과를 검증하는 데 중요한 역할
    - 여러 독립변수가 종속변수에 미치는 영향을 동시에 평가
- 활용
    - 의학 연구
    - 소비자 행동 연구에서 다양한 제품 유형 간 선호도 비교
    - 제조업에서 다양한 생산 공정의 효과를 비교
### 8.1.3 분산분석의 기본 가정
1. 독립성
- 각 그룹의 표본들은 독립적으로 추출되어야 함
- 실험 또는 조사에서 데이터 수집 과정에서의 독립성 의미
2. 정규성 Normality
- 각 그룹의 데이터가 정규 분포를 따라야 함
    - 데이터가 정규 분포를 따르지 않으면 결과가 왜곡될 수 있음
- 특히, 그룹 표본 크기가 작을 때 중요
3. 등분산성 Gomoscedasticty
- 각 그룹의 분산이 동일해야 함
- 그룹 간 데이터 변도이 유사해야 한다는 의미
- 만약 그룹 간 분산이 다르면, 분산분석의 결과를 신뢰할 수 없음

## 8.2 일원 분산분석 One-way ANOVA
### 8.2.1 일원 분산분석의 개념
- 하나의 독립 변수가 여러 그룹에 미치는 영향
- '일원'은 분석에 하나의 독립 변수만 사용된다는 뜻
### 8.2.2 그룹 간 차이 분석
- 총 분산 Total Variation
    - 전체 데이터의 변동
    - 그룹 간 분산과 그룹 내 분산의 분해
- F-통계량
    - 그룹 간 분산이 그룹 내 분산에 비해 얼마나 큰지를 나타내는 비율
    - f-통계량이 클수록 그룹 간 평균의 차이가 큼

    

## 8.3 이원 분산분석 Two-way ANOVA)
### 8.3.1 이원 분산분석의 개념
- 두 개의 독립 변수가 종속 변수에 미치는 영향을 동시에 분석하는 방법
- 각 독립 변수의 주효과와 두 독립 변수 간 상호작용 효과도 분석 가능
- 주효과 Main Effect
    -  각 독립 변수가 종속 변수에 미치는 독립적인 영향
- 상호작용 효과 Interacion Effect
    - 한 독립 변수 효과가 다른 독립 변수의 수준에 따라 달라지는 효과
### 8.3.2 상호작용 효과 분석
- 한 독립 변수의 효과가 다른 독립 변수의 수준에 따라 달라지느지를 보여줌
- 상호작용 효과가 유의미
    - 두 독립 변수가 종속 변수에 미치는 영향이 독립적으로 작용하지 않음
    - 한 요의 효과가 다른 요의 수준에 따라 변함
- 상호작용 효과가 유의미하지 않음
    - 두 독립 변수는 종속 변수에 독립적으로 영향을 미침
    - 서로의 영향을 받지 않음

## 8.4 반복 측정 분산분석 Repeated Measures ANOVA
### 8.4.1 반복 측정 분산분석의 개념
- 같은 피험자나 실험 대상이 여러 번 측정되는 경우, 그 결과들 사이의 차이를 분석하기 위한 통계 기법
- 피험자가 각기 다른 조건에서 반복적으로 측정될 때, 조건 간 차이를 확인하는 데 유용
- 반복 측정 분산분석은 각 피험자 자신의 대조군 역햘을 하기 때문에 피험자 간 변동을 통제할 수 있어 더욱 정교한 분석 가능
- 예시
    - 어떤 약물이 시간 경과에 따라 혈압에 미치는 영향을 평가하는 연구에서 동일한 환자의 혈압으로 약물 복용 전, 1시간 후, 2시간 후에 측정하는 경우
### 8.4.2 반복 측정 디자인의 이해
- 같은 피험자나 대상에게 여러 번 측정을 수행하는 실험 디자인
- 이 디자인에서는 동일한 피험자들이 여러 조거이나 시간점에서 반복적으로 측정
- 피험자 간 변동을 최소화하여 각 조건 간 차이를 더 명확하게 분석 가능
- 피험자 간 변동
    - 피험자 개개인의 고유한 특성으로 인한 변동
    - 반복 측정에서는 이 변동을 통제 가능

## 8.5 공분산분석 Analysis of Covariance ; ANCOVA
- 분산분석과 회귀분석을 결합한 통계기법
- 그룹 간 차이를 분석할 때 공변량covariate을 고려하여 분석의 정확성을 높이는 방법
- 그룹 간 차이를 비교할 때 공변량에 의한 영향을 통제하여 순수한 그룹 효과를 평가
- 그룹 간 평균 차이를 비교할 때 다른 변수(공변량, covariate)가 종속 변수에 미치는 영향을 통제하는 방법
- 독립 변수와 종속 변수 사이의 관계를 분석하는 ANOVA에 공변량을 추가하여 회귀분석의 개념을 적용한 것
- 분석에서 제거하고자 하는 외부 요인의 영향을 줄이기 위해 공변량 효과 통제
- 공변량을 통제 후 그룹 간 차이가 여전히 유의미한지 평가
### 8.5.2 공변량의 역할과 해석
- 공변량 covariate
    - 종속 변수에 영향을 미칠 수 있지만 연구의 주요 관심사가 아닌 변수
- 분석에서 통제하여, 주효과(독립 변수의 효과)를 더 정확하게 평가하는 역할
- 공변량이 제거되면, 주효과가 더 명확히 드러나 혼동 요인 제거 효과가 있음
- 공변량을 통제함으로써 오류 변동을 줄이고, 분석의 정밀도를 높일 수 있음
- 공변량이 유의미하며, 종속 변수에 실질적인 영향을 미친다는 것을 의미
    - 이때, ANCOVA는 공변량의 효과를 제거한 후에도 독립 변수 간 차이가 유의미한지 평가
- 공변량이 통제된 상태에서도 그룹 간 차이가 여전히 유의미하다면, 독립 변수의 효과가 공변량의 영향을 배제하고도 존재한다고 해석 가능

## 8.6 다중 비교 방법
- 다중 비교 Multiple Comparisons
    - 여러 그룹 간 평균 차이를 동시에 비교하는 과정에서 발생하는 문제를 해결하기 위한 통계적 방법
    - 분산분석에서 그룹 간 차이가 유의미하다고 결론이 나면, 구체적으로 어떤 그룹들 간 차이가 있는지 알아보기 위해 다중 비교 방법을 사용
    - 다중 비교는 여러 번의 비교를 수행함으로써 발생할 수 있는 오류를 통제하기 위해 설계
### 8.6.1 사후검정 Post Hoc Tests
- ANOVA에서 그룹 간 차이가 유의미하다고 결론 내려진 후. 어떤 그룹들 간 실제로 차이가 있는지 확인하기 위한 추가 분석
- ANOVA자체는 그룹 간 평균의 전체적일 차이만 평가하기 때문에 개별 그룹 간 차이를 명확히 식별하기 위해 사후검정 필요
- 사후 검정 방법
    - Tukey's HSD ; Honestly Significant Difference
    - Bonferroni 보정
    - Scheffé 방법
- 사후검정은 다중 비교로 인해 발생할 수 있는 제1종 오류를 제어하면서, 그룹 간 차이를 확인 가능
- 예시
    - 세가지 교육 방법이 학생 성적에 미치는 영향을 분석 후 각 교육 방법 간 성적 차이를 비교하기 위해 사후검정 사용
### 8.6.2 다중 비교 문제와 보정 방법
- 다중 비교 문제는 여러 그룹 간 평균을 비교할 때, 제1종 오류가 증가하는 문제
- 여러 번의 비교를 동시에 수행하면, 유의수준α이 누적되어 잘못된 결론에 도달할 확률이 증가
- 이 문제를 해결하기 위해 다중 비교 보정 방법 사용
    - 유의수준을 낮춰 다중 비교로 인해 증가할 수 있는 오류 제어
    - 다중 비교에 대한 엄격한 기준을 설정해 잘못된 긍정 오류(제1종 오류) 가능성 낮춤
- Bonferroni 보정
    - 비교의 수만큼 유의수준을 나눠 적용
    - 예시
        - 3가지 비교를 수행할 때 유의수준 α = 0.05라면, 각각 0.0167로 설정
- FDR 보정 ; False Discovery Rate
    - 허위 발견율을 조정하여 다중 비교에서 발생할 수 있는 오류를 줄임
### 8.6.3 Tukey's HSD, Bonferroni, Scheffé 방법
- Tukey's HSD ; Honestly Significant Difference
    - 모든 가능한 그룹 쌍 간 평균 차이를 비교하는 방법
    - 비교의 수에 따라 유의수준을 조정해 제1종 오류를 제어
    - 동일한 표본 크기와 유사한 분산을 가정할 때 사용
    - 결과로 도출된 HSD값을 기준으로 각 그룹 간 차이가 유의미한지 판단
- Bonferroni 보정
    - 가장 간단한 다중 비교 보정 방법
    - 유의수준을 비교의 수로 나눠 제1종 오류 낮춤
    - 비교의 수가 적을 때 효과적
    - 보수적인 방법으로 알려져 있음
    - 보장된 유의수준에 따라 비교의 유의성 평가
- Scheffé 방법
    - 다중 비교를 보수적으로 접근하는 방법
    - 모든 가능한 비교를 포함하여 분석
    - 유의수준을 크게 조정해 제1종 오류를 매우 엄격하게 제어
    - 여러 비교를 수행하는 복잡한 분석에서 주로 사용
    - 다른 방버베 비해 유연성이 높은 대신 보수적
    - 비교 결과를 매우 보수적으로 평가하므로, 유의미한 차이가 도출되기 어려움
    - 대신 신뢰성 높은 결론

## 8.7 분산분석의 고급 주제
- ANOVA는 몇 가지 가정을 따르며, 이 가정에 위배되거나 더 복잡 데이터 구조가 있는 경우, 고급 주제를 다루는 기술이 필요
### 8.7.1 이질적 분산을 고려한 분석
- 이질적 분산 Heteroscedasticity
    - 각 그룹의 분산이 동일하지 않을 때 발생하는 문제
- 기본적으로 ANOVA는 모든 그룹의 분산이 동일하다는 가정을 하지만 현실에서 그룹 간 분산이 다를 수 있음
- Welch’s ANOVA
    - 그룹 간 분산이 동일하지 않을 때 사용하는 방법
    - 각 그룹 분산을 고려하여 F-통계를 조정
- Brown-Forsythe 검정
    - 그룹 간 중위수를 사용해 분산이 동일하지 않은 경우에도 유요한 분석 결과 도출 가능
### 8.7.2 다변량 분산분석 MANOVA ; Multivariate Analysis of Variance
- 여러 종속 변수들에 대해 그룹 간 차이를 동시에 분석하는 방법
- 기본적인 ANOVA는 단일 종속 변수에 대해서만 분석하지만, MANOVA는 다수의 종속 변수를 포함하여 그들 간 상호 관계 분석 가능
- 종속 변수들이 상호 연관되어 있을 때 유용
- 여러 종속 변수 간의 상관관계를 고려하여 분석하므로, 보다 통합된 결론 도출 가능
- 여러 종속 변수가 있는 고차원 데이터를 다룰 때 유용
### 8.7.3 혼합 설계 분산분석 Mixed-Design ANOVA
- 독립 변수가 반복 측정 요인과 독립 요인으로 구성된 경우 사용하는 방법
- 혼합 설계에서는 일부 요인은 각 실험 단위 내에서 반복 측정되고, 다른 요인은 독립된 상태에서 측정
- 복잡한 실험 디자인을 분석할 때 유용
- 반복 측정된 요인과 독립된 요인 간 상호작용 분석 가능
- 실험 설계가 복잡할 수록 혼합 설계 분산분석이 유용
- 예시
    - 교육 방법이 시간에 따라 학생들의 성적에 미치는 영향을 분석할 때, 각 학생의 성적은 반복적으로 측정되고, 교육방법은 독립된 요인
