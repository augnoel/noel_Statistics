# 12. 베이즈 정리 Bayes' Theorem

## 12.1 베이즈 정리의 기초
### 12.1.1 베이즈 정리의 정의
- 주어진 데이터에 대해 어떤 사건이 발생할 확률을 갱신하는 방법을 제공



- 사전 확률 P(A)을 새로운 증거 B에 따라 수정하여 사후 확률 P(A|B)을 계산하는 방식으로 작동
### 12.1.2 조건부 확률과 베이즈 정리의 관계
- 조건부 확률은 두 사건이 연관되어 있을 때 한 사건이 발생했을 때 다른 사건이 발생할 확률을 의미



- 사건 B에 대한 조건부 확률을 기반으로 사건 A의 사후 확률을 계산하는 것
- 베이즈 정리르 통해 초기 사전 확률을 데이터나 증거를 통해 업뎃 가능

## 12.2 베이즈 정리의 수학적 표현
### 12.2.1 베이즈 정리의 수식 도출
- 기본적인 조건부 확률



- 여기서 A와 B가 동시에 발생할 확률 식은



- 사건 B가 발생했을 때, 사건 A가 발생할 확률



- 위의 식들을 A와 B가 동시에 발생할 확률에 관해 정리하면,



- P(A∩B)에 관해 정리하면



- B가 발생할 확률로 나눠 정리하면 베이즈 정리 도출
- 주어진 사건 B에 대해 사건 A의 사후 확률을 계산




### 12.2.2 사전 확률, 사후 확률, 우도 함수의 개념
- 사전 확률 Prior Probability, P(A)
    - 사건 A가 발생할 확률
    - 새로운 증거 B가 도입되기 전에 알고 있거나 가정한 확률
    - 사전 확률은 과거의 경험이나 기존의 지식을 기반으로 설정
- 사후 확률 Posterior Probability, P(A|B)
    - 사건 B가 발생한 후 사건 A가 발생할 확률
    - 새로운 증거 B를 반영하여 사전 확률을 업뎃한 결과
    - 베이즈 정리의 핵심(사후 확률을 계산하는 것)
- 우도 함수 Likelihood, P(B|A)
    - 사건 A가 발생한 경우, 사건 B가 발생할 조건부 확률
    - 관찰된 데이터가 주어진 가설A 하에서 얼마나 일어날 가능성이 있는지 나타냄
### 12.2.3 베이즈 정리와 확률 갱신
- 확률 갱신 Probability Updating
    - 새로운 증거가 제시될 때 기존의 사전 확률을 수정하여 더 정확한 사후 확률을 추정하는 과정
- 과정
1. 사전 확률
    - 처음 사건 A에 대한 사전 확률이 주어짐
    - 과거의 경험, 데이터, 주관적인 판단에 의해 설정
2. 우도 
    - 주어진 사건 A가 발생했을 때, 사건 B가 발생할 확률
    - 관찰된 데이터가 주어진 가설을 얼마나 잘 설명하는지 나타냄
3. 증거 Evidence, P(B)
    - 사건 B가 발생할 전체 확률을 계산
    - 모든 가능한 사건이 발생할 확률을 반영
4. 사후 확률
    - 베이즈 정리를 사용해 사건 B가 발생 후 사건 A의 사후 확률을 계산
    - 사전 확률은 새로운 증거에 따라 갱신한 결과

## 12.3 베이즈 정리의 적용
### 12.3.1 베이즈 정리의 활용
- 나이브 베이즈 분류기
    - 베이즈 정리를 기반으로 하는 기계 학습 알고리즘
    - 데이터의 특정 특징들이 독립적이라고 가정
    - 이 특징들에 대한 사후 확률을 계산하여 데이터 분류
- 의학에서의 질병진단, 법의학의 증거 평가, 금융에서 리스크 관리에 사용됨
### 12.3.2 베이즈 정리를 사용한 추론 방법
1. 사전 확률 설정
- 사건에 대한 초기 믿음, 즉 사전 확률을 설정
- 과거 데이터, 경험, 주관적 판단에 기반
2. 우도 계산
- 새로운 데이터나 증거가 주어졌을 때, 이 증거가 주어진 가설 하에서 얼마나 발생할 가능성이 있는지 나타냄
3. 사후 확률 계산
- 베이즈 정리르 사용하여 사전 확률을 새로운 증거에 따라 갱신하고, 이를 통해 사후 확률을 계산
- 새로운 데이터를 반영한 최신의 믿음
4. 결정 및 행동
- 계산된 사후 확률을 바탕으로 의사결정
### 12.3.3 베이즈 정리와 의사결정
- 정보 갱신
    - 새로운 정보가 들어올 떄마다 기존의 확률 갱신함으로써, 의사결정자가 현재 가장 신뢰할 수 있는 정보를 바탕으로 결정을 내릴 수 있게 함
- 비용-이익 분석
    - 베이즈 정리를 통해 사후 확률을 계산하고, 이를 바탕으로 다양한 선택지의 기대 비용과 이익을 비교하여 최적의 결정
- 리스크 평가
    - 의사결정 과정에서의 리스크를 평가할 때, 베이즈 정리를 불확실성의 수준을 수량화하는 데 도움을 줌
    - 재정적 결정이나 전략적 계획 수립에 유용
- 상황 변화에 대한 적응
    - 베이즈 접근법은 변화하는 상황에 따라 신속하게 확률 조정 가능
    - 의사결정자가 새로운 상황에 빠르게 적응할 수 있게 함

## 12.4 베이즈 정리와 통계학
### 12.4.1 베이지안 통계와 빈도주의 통계의 비교
- 확률 해석
    - 빈도주의 통계
        - 확률을 장기적인 빈도로 해석
        - 예시, 동전을 던졌을 때 앞면이 나올 확률이 0.5라는 것은 무한히 반복되는 실험에서 앞면이 나올 확률이 50%라는 것
    - 베이지안 통계
        - 확률을 주관적인 믿음으로 해석
        - 확률은 특정 사건이 일어날 가능성에 대한 개인적인 신뢰도
        - 예시, 앞면이 나올 확률 0.5는  현재의 정보와 믿음에 기반해 사건이 일어날 가능성을 평가한 것
- 가설 검정과 추론
    - 빈도주의 통계
        - 귀무가설을 설정하고, 데이터를 바탕으로 이 가설을 기각할지 말지를 결정
        - p-값은 귀무가설이 참이라는 가정 하에 데이터를 얻을 확률
    - 베이지안 통계
        - 사전 확률을 설정한 후, 데이터를 바탕으로 사후 확률 계산
        - 데이터를 통해 가설이 참일 확률을 직접 계산하고, 이 확률에 기반해 결정
- 데이터 활용
    - 빈도주의 통계
        - 일반적으로 고정된 데이터를 바탕으로 추론 수행
        - 새로운 데이터가 추가되면 기존 추론 결과와 별도로 다시 분석
    - 베이지안 통계
        - 기존 데이터와 새로운 데이터를 결합하여 사후 확률 갱신 가능
        - 동적인 데이터 분석과 지속적인 정보 갱신에 유리
### 12.4.2 베이지안 추정과 신뢰 구간의 차이
- 베이지안 추정과 빈도주의의 신뢰 구간은 모두 추정값의 불확실성을 나타냄
- 신뢰 구간 Frequentist Confidence Interval
    - 빈도주의 통계에서 특정 매개변수가 포함될 것으로 예상되는 범위
    - 예시, 95% 신뢰 구간은 동일한 실험을 여러 번 반복했을 때 그 중 95%에서 참된 매개변수가 구간 내에 있을 것이라는 의미
    - 이는 반복적인 실험과 장기적인 빈도에 기반
- 베이지안 추정 Bayesian Estimation
    - 매개변수의 추정치는 사후 확률 분포로 나타냄
    - 이 분포를 바탕으로 특정 매개변수가 가질 확률 범위를 계산
    - 예시, 95% 베이지안 신뢰 구간은 주어진 데이터와 사전 확률을 고려할 때, 매개변수가 이 구간 내에 있을 확률이 95%
    - 이는 데이터와 사전 정보에 기반한 개인적 믿음의 정도
### 12.4.3 베이지안 모델링의 장단점
- 장점
1. 사전 정보 활용
- 사전 확률을 통해 기존의 지식이나 정보를 통합 가능
- 경험적 데이터가 부족할 때 유리
- 서전 정보를 반영한 보다 정교한 추론 가능
2. 불확실성의 직관적 표현
- 불확실성을 확률로 직접 표현
- 다양한 상황에서 불확실성에 대한 명확한 해석 제공 가능
3. 동적 갱신
- 새로운 데이터가 추가될 때마다 사후 확률 갱신 가능
- 변화하는 상황에 신속하게 적응 가능
- 점진적인 학습 및 데이터 통합에 유리
4. 복잡한 모델 적용 가능
- 복잡한 모델을 다룰 때도 유연하게 적용 가능
- 다양한 추론을 동시에 수행 가능
- 단점
1. 계산 복잡성
- 종종 계산적으로 매우 복잡
- 특히 고차원 매개변수 공간에서 사후 분포를 계산하는 데 상당한 컴퓨팅 자원 필요
- 이를 해결하기 위해 마코프 연쇄 몬테카를로MCMC 방법과 같은 복잡한 계산 기법 필요
2. 사전 확률의 주관성
- 사전 확률의 설정이 주관적일 수 있음
- 잘못 설정된 사전 확률은 결과에 큰 영향
- 사전 확률 설정은 신중하게 진행해야 함
3. 해석의 난이도
- 빈도주의 접근법에 비해 더 많은 통계적 배경 지식이 필요
- 사전 확률과 사후 확률의 개념이 익숙하지 않은 사람들에게 직관적으로 이해하기 어려움

## 12.5 베이즈 정리의 확장
### 12.5.1 베이즈 정리의 다차원 확장
- 베이즈 정리는 기본적으로 두 사건 A, B 사이의 관계를 설명
- 다차원으로 확장하여 이려 사건 간의 관계 분석도 가능
- 다변량 베이즈 정리multi-dimensional Bayesian theorem가 사용
- 여러 변수들의 조합에 대한 조건부 확률 계산
- 다차원 베이즈 정리
    - 여러 변수 X_1, X_2 ~ X_n이 주어질 때, 특정 사건 Y가 발생할 확률을 계산

    

    - 다수의 조건부 확률을 결합하여, 전체 사건의 발생 확률 추정
    - 복잡한 시스템이나 데이터에서 여러 요인이 결합된 상황을 분석
### 12.5.2 베이즈 네트워크 Bayesian Networks
- 확률적 모델로 여러 변수 간 상화 의존성을 그래프로 변환
- 복잡한 확률 관계를 시각적으로 표현하고, 계산을 효율적으로 수행하게 해줌
- 각 노드는 변수(사건)을 나타냄
- 노드 간 연결선은 조건부 의존 관계을 상징
- 노드 Node
    - 네트워크에서의 변수 또는 사건
- 엣지 Edge
    - 노든 간 관계를 나타내며, 방향성이 있는 그래프
    - 방향은 원인에서 결과로의 흐름
- 조건부 확률표 Conditional Probability Table ; CPT
    - 각 노드는 자신에게 들어오는 부모 노드들에 대해 조건부 확률을 가짐
    - 실험이나 데이터, 사전 지식에 기반해 설정
- 장점
    - 복잡한 시스템의 구조적 이해를 돕고, 다양한 확률적 의사결정 지원
    - 주어진 증거, 관찰된 데이터를 바탕으로 다른 변수들의 확률을 업뎃, 추론 가능
    - 부분적 정보만을 이용해 추론 가능
    - 시스템의 불확실성을 효과적으로 관리 
### 12.5.3 마코프 연쇄 몬테카를로MCMC 방법과 베이즈 추론
- 복잡한 확률 분포를 샘플링하기 위해 사용되는 계산 기법
- 베이즈 모델의 사후 확률 분포를 정확하게 계산하기 어려운 경우, MCMC를 통해 근사적인 해결책 모색 가능
- 마코프 연쇄 Markov Chain
    - 현재 상태가 이전 상태에만 의존하는 확률적 과정
    - 이전 상태의 정보만으로 다음 상태의 확률 결정 가능
- 몬테카를로Monte Carlo 방법
    - 확률적 샘플링 기법
    - 복잡한 수학적 문제를 근사적으로 해결
- 베이즈 모델에서 사후 확률 분포를 계산할 때, MCMC를 통해 이 분포에서의 샘플을 생성하고, 이를 기반으로 추론
- Metropolis-Hastings 알고리즘과 Gibbs 샘플링이 있음
- 복잡한 고차원 문제에서 베이즈 추론을 효과적으로 수행 가능
- 장점
    - 복잡한 분포에서의 직접적인 확률 계산이 어려윤 경우, 근사적인 샘플링을 통해 추론 가능
    - 다양한 베이즈 모델에서 유연하게 적용 가능
    - 고차원 문제에서도 효과적으로 작동
