# 09 회귀 분석의 통계적 기초

## 9.1 회귀 분석의 개요
### 9.1.1 회귀 분석의 정의
- 회귀 분석 Regression Analysis
    - 종속 변수와 하나 이상의 독립 변수 간 관계를 모델링하고 분석하는 통계 기법
    - 주어진 데이터에서 패턴을 찾아내어 예측 모델을 구축하거나 변수 간 상관 관계를 이해하는 데 사용
- 특정 독립 변수가 종속 변수에 미치는 영향은 얼마나 되는가?
- 여러 독립 변수의 조합이 종속 변수에 어떤 영향을 미치는가?
- 주어진 독립 변수 값으로 종속 변수를 예측할 수 있는가?
- 회귀 분석은 선형 회귀 분석과 비선형 회귀 분석으로 나뉨
### 9.1.2 회귀 분석의 목적
1. 예측 Prediction
- 과거 데이터를 기반으로 미래의 종속 변수 값 예측
2. 변수 간 관계 이해 Understanding Relationships
- 독립 변수들이 종속 변수에 어떻게 영향을 미치는지 분석하여 변수들 간 관계를 이해
3. 모델링과 해석 Modeling and Interpretation
- 회귀 분석을 통해 데이터의 구조와 특성을 모델링하고. 모델의 계수 값을 해석하여 각 독립 변수가 종속 변수에 미치는 영향력 설명
4. 통제 변수의 효과 제거 Controlling for Variables
- 다중 회귀 분석을 사용하면 여러 변수의 영향을 동시에 고려할 수 있어 특정 변수의 순수한 효과를 추정 가능
### 9.1.3 회귀 분석의 기본 가정
1. 선형성 가정 Linearity
- 독립 변수와 종속 변수 간 관계가 선형이어야 함
- 변수들이 직선 형태의 관계를 갖는다고 가정
2. 독립성 가정 Independece
- 관측값들은 서로 독립적
- 특히, 잔차(오차)가 독립적이어야 하며 이는 시간이나 순서에 따른 데이터의 경우 매우 중요
3. 등분산성 가정 Homoscedasticity
- 모든 수준의 독립 변수에 대해 잔차의 분사이 동일
- 예측된 종속 변수 값에 상관없이 잔차의 분포가 일정해야 함
4. 정규성 가정 Normality
- 잔차가 정규 분포를 따름
- 회귀 모델의 신뢰 구간이나 가설 검정에서 중요
5. 다중 공선성 없음 No Multicollinearity
- 다중 회귀 분석에서 독립 변수들 간 강한 상관 관계가 없어야 함
- 다중 공선성이 있으면 회귀 계수의 불안정과 해석의 어려움이 발생할 수 있음

## 9.2 선형 회귀 분석
- 종속 변수와 하나 이상의 독립 변수 간의 관계를 직선 형태로 모델링
- 예측과 변수 간의 관계를 파악하는데 유용
### 9.2.1 단순 선형 회귀 Simple Linear Regression
- 하나의 독립 변수와 하나의 종속 변수 사이의 관계를 분석



- x와 y간의 관계가 선형인지 관계 분석
- x를 알 때 y값 예측
- 회귀 계수 β_0, β_1은 주어진 데이터에 맞춰 최소 제곱법Ordinary Least Squares ; OLS을 사용해서 추정
- OLS는 잔차 제곱 합을 최소화하는 방식으로 최적의 회귀선 찾음 
### 9.2.2 다중 선형 회귀 Multiple Linear Regression
- 여러 독립 변수를 사용해 하나의 종속 변수를 예측하는 방법



- 여러 변수들이 종속 변수에 어떻게 영향을 미치는지 다중 변수의 영향력 분석
- 여러 독립 변수를 고려해 종속 변수를 더 정확하게 예측
- 단순 선형 회귀와 동일하게 OLS 방법을 이용해 회귀 계수 추정
### 9.2.3 회귀 계수의 해석
- 회귀 계수 β_i의 의미
    - 각 회귀 계수β_i는 독립 변수 x_i가 한 단위 증가할 때 종속 변수 y에 미치는 평균적인 변화를 나타냄
    - 예시
        - β_1 = 2일때, x_1이 1단위 증가할 때마다 y가 평균적으로 2 단위씩 증가함
- 절편 β_0
    - 모든 독립 변수 x_i = 0일 때, 예측된 종속 변수 y값
    - 주로 이론적으로 의미를 가짐
    - 독립 변수가 0일 가능성이 없는 경우에도 모델에서 중요한 역할을 함
- 유의성 검정
    - 각 회귀 계수의 유의성을 검정하여 해당 독립 변수가 종속 변수에 통계저긍로 유의미한 영향을 미치는 판판
    - p-값을 사용하며, p-값 < 유의수준 일때, 해당 변수의 회귀 계수는 통계적으로 유의미하다고 할 수 있음
- 다중 공선성
    - 다중 선형 회귀에서 독립 변수들 간 상관관계가 높을 경우, 회귀 계수의 해석이 어려워 질 수 있음
    - 이를 해결하기 위해 VIF ; Variance Inflation Factor와 같은 지표를 사용해 공선성을 진단하고 필요한 경우 변수를 조정해야 함

## 9.3 회귀 모델의 적합도 평가
- 회귀 분석을 수행 후 모델의 성능과 적합도를 평가해야 함
- 모델이 데이터를 얼마나 잘 설명하는지, 예측의 정확성이 얼마나 높은지를 판단
### 9.3.1 결정계수R²와 조정된 결정계수
- 결정계수 R²
    - 회귀 모델이 종속 변수의 변동을 얼마나 설명하는지 나타내는 지표
    - 0 ~ 1사이의 값을 가지며, 1에 가까울수록 모델이 데이터를 잘 설명함을 의미

    
- 조정된 결정계수 Adjusted R²
    - R²의 단점을 보완한 지표
    - 변수의 수에 따른 모델의 과적합 방지를 위해 사용
    - 독립 변수가 많아질수록 R²는 증가하는 경향이 있는데 조정된 R²는 모델의 복잡성을 반영하여 보정된 값을 제공

    

### 9.3.2 잔차 분석 Residual Analysis
- 잔차 Residuals
    - 각 관찰값에서 예측값을 뺀 차이
    - 모델이 설명하지 못한 부분을 의미
    - 잔차 분석은 이 잔차들이 모델 가정에 부합하는 평가하는 과정
- 잔차의 평균은 0 
    - 이는 모델이 편향되지 않았음을 의미
- 정규성을 따름
    -  잔차들이 정규 분포를 따라야 함
    - 이를 확인하기 위해 Q-Q 플롯을 사용하거나 샤피로-윌크 검정 등을 수행 가능
- 등분산성 Homoscedasticity 
    - 잔차의 분산이 일정해야 함
    - 독립 변수의 값에 관계 없이 잔차의 분산이 일정해야 함
    - 만약 분산이 일정하지 않다면 이분산성Heteroscedasticity 문제가 있을 수 있음
- 자기상관이 없어야 함
    - 잔차들 간 상관관계가 없어야 함
    - 잔차가 독립적이어야 함
    - 자기상관이 존재하는 경우, 모델이 독립성을 위반한 것
- 잔차 플롯
    - 잔차를 x축, 예측값 또는 독립 변수를 y축에 둔 그래프
    - 패턴이 잘 보이지 않으면 모델이 잘 맞는 것
### 9.3.3 모델의 유의성 검정(F-검정)
- F-검정
    - 회귀 모델 전체가 통계적으로 유의미한지를 평가하는 방법
    - 회귀 모델이 유의미한 예측력을 가지는지, 독립 변수들이 종속 변수에 실질적인 영향을 미치는지 검증
    - 귀무가설 H_0
        - 모든 회귀 계수 β_i는 0
        - 독립 변수 들이 종속 변수에 영향을 미치지 않음
    - 대립가설 H_1
        - 최소한 하나의 회귀 계수는 0이 아님
        - 독립 변수 중 최소 하나가 종속 변수에 유의미한 영향을 끼침
- F-통계량



## 9.4 회귀 분석의 가정과 진단
- 회귀 분석을 수행할 때, 결과의 신뢰성과 정확성을 보장하기 위해 몇 가지 기본적인 가정을 충족해야 함
- 이 가정들은 모델의 적합성을 평가하고, 회귀 계수의 추정치가 편향되지 않고 신뢰할 수 있는지 확인
### 9.4.1 선형성 Linearity
- 종속 변수와 독립 변수 간 관계가 선형적이어야 한다는 가정
- 독립 변수의 변화가 종속 변수의 변화에 일정한 영향을 미침
- 잔차 플롯을 이용해 잔차가 무작위로 분포하고 특정 패턴이 모이지 않는다면 선형성 가정이 충족
- 잔차 플롯에서 패턴이 존재해 비선형 관계를 시사한다면 변수 변환(로그 변환, 제곱근 변환)을 시도하거나 다항 회귀 같은 비선형 회귀 모델 고려
### 9.4.2 독립성 Independence
- 회귀 모델의 잔차들 간에 자기상관이 없다는 가정을 의미
- 잔차가 서록 독립적이어야 함
- 더빈-왓슨 통계량 Durbin-Watson statistic
    - 잔차들 간의 자기상관 검사 가능
    - 더빈-왓슨 값이 2에 가까울수록 자기상관이 없음
    - 값이 0 또는 4에 가까울수록 양의 자기상관이나 음의 자기 상관이 있을 수 있음
- 자기상관이 발견되면 시간의 흐름에 따라 데이터가 수집된 경우 시계열 분석 기법을 사용하거나 독립 변수로 시간 변수를 추가하는 방법 등을 고려
### 9.4.3 등분산성 Homoscedasticity
- 모든 독립 변수 값에 대해 잔차의 분산이 동일해야 한다는 가정
- 독립 변수의 크기에 상관없이 잔차의 변화 범위가 일정해야 함
- 등분산성이 충족된다면 잔차 플롯에서 잔차가 무작위 분포
- 만약 잔차가 깔때기 모양으로 퍼지거나 좁아진다면 이분산성이 존재할 수 있음
- 이분산성이 발견될 경우 가중 회귀weighted regression나 로그 변화을 통해 문제 완화
### 9.4.4 정규성 Normality
- 잔차가 정규 분포를 따라야 한다는 가정
- 회귀 계수의 추론과 p-값 계산이 올바르게 이루어지기 위해 중요
- Q-Q플롯 
    - 잔차의 정규성 확인을 위해 사용
    - 잔차가 직선에 가깝게 분포할소록 정규성 가정이 충족된 것
- 샤피로-윌크 검정 or 콜모고로프-스미르노프 검정
    - 잔차가 정규 분포를 따르는지 통계적으로 검정 가능
- 잔차가 정규 분포를 따르지 않을 경우 로그 변환이나 Box-Cox 변환 등으로 데이터를 변환하거나 비모수 회귀 방법 고려

## 9.5 다중 공선성과 변수 선택
- 회귀 분석에서 독립 변수들이 상호 독립적이야 한다는 가정이 중요한데, 이 가정이 위배되면 다중 공선성Multicollinearity 문제가 발생
- 다중 공선성은 회귀 분석의 결과를 왜곡시킬 수 있음
### 9.5.1 다중 공선성 문제
- 2개 이상의 독립 변수들이 강한 상관관계를 가질 때 발생
- 각 독립 변수가 종속 변수에 미치는 개별 효과를 분리해내기 어려움
- 회귀 계수의 불안정성 및 해석의 어려움 초래
- 계수의 불안정
    - 다중 공선성이 존재하면 회귀 계수의 추정치가 매우 민감
    - 데이터의 작은 변화에도 큰 영향을 받을 수 있음
- 해석의 어려움
    - 독립 변수 간의 상관관계가 높기 때문에, 각각의 독립 변수가 종속 변수에 미치는 영향을 명확히 구분하기 어려움
- 모델의 신뢰성 저항
    - 다중 공선성이 심한 경우, 모델의 예측력과 신뢰성이 떨어질 수 있음
- 분산 팽창 계수 VIF ; Variance Inflation Factor
    - VIF를 계산하여 다중 공선성 평가
    - 10이 넘으면 심각한 다중 공선성이 존재한다고 판단
- 조건수 Condition Number
    - 다중 공선성 유무 평가
    - 30을 초과하면 다중 공선성이 존재할 가능성이 있다고 판단
### 9.5.2 변수 선택 방법(전진 선택법, 후진 제거법, 단계적 회귀)
- 회귀 분석에서 다중 공선성 완화하거나, 모델의 예측력을 향상시키기 위해 불필요한 독립 변수를 제거하거나 적절한 변수를 선택하는 과정
1. 전진 선택법 Forward Selection
- 초기 모델에 변수가 하나도 없는 상태에서 시작하여 통계적으로 유의미한 변수를 하나씩 추가하는 방식
- 가장 유의미한 변수를 먼저 추가하고, 이후 단계별로 모델의 적합도를 평가하며 변수 추가
- 단순하고 해석이 쉬우며, 중요한 변수를 선택하는 데 효과적
- 이미 포함된 변수에 의해 나중에 초가될 변수의 중요성이 과소평가될 수 있음
2. 후진 제거법 Backward Elimination
- 초기 모델에 독립 변수를 포함시키고, 통계적으로 유의미하지 않음 변수를 하나씩 제거하는 방법
- 가장 유의미하지 않은 변수부터 제거하면서 모델의 적합도 평가
- 처음부터 모든 변수를 고려하므로 중요한 변수를 빠뜨릴 가능성이 낮음
- 계산 비용이 높고 다중 공선성이 심한 경우 해석이 어려움
3. 단계적 회귀 Stepwise Regression
- 전진 선택법과 후진 제거법을 결합한 방식
- 변수 추가와 제거를 동시에 고려하여 최적의 변수 조합을 찾음
- 한 변수를 추가 후 그 변수가 유의미한지 평가하고 필요에 따라 제거
- 변수의 추가와 제거를 동적으로 수행하여 과적합 방지와 최적의 모델 찾기에 효율적
- 모델이 복잡해질 수 있고, 데이터에 민감하게 반응
### 9.5.3 VIF ; Variance Inflation Factor 계산 및 해석
- 다중 공선성을 진단하는 지표
- 특정 독립 변수가 다른 독립 변수들과 얼마나 상관되어 있는지 나타냄
- VIF가 클수록 다중 공선성이 큼



- 다중 공선성을 해결하기 위해
    - 독립 변수 제거
        - VIF 값이 높은 변수를 모델에서 제거하여 다중 공선성 완화
    - 주성분 분석 PCA
        - 독립 변수들의 선형 결합을 통해 새로운 변수를 만들어 다중 공선성 낮출 수 있음
    - 정규화 Regularization
        - 릿지 회귀와 같은 정규화 기법을 사용하여 다중 공선성 영향을 낮출 수 있음
