# 04. 가설 검정 Hypothesis Testing
## 4.1 가설 검정의 기초
- 통계학에서 특정 주장이나 가정을 검증하기 위한 방법
- 연구자는 2가지 가설을 세우고, 데이터 분석을 통해 어떤 가설이 더 타당한지 평가

### 4.1.1 가설의 종류 (귀무가설과 대립가설)
1. 귀무가설 Null Hypothesis, H_0
- 연구에서 검증하고자 하는 기본 가정
- 보통 효과 없다 or 차이가 없다는 주장을 포함
- 연구자는 이 가설을 기각할 수 있는지 여부를 판단
2. 대립가설 Alternative Hypothesis, H_1 or H_A
- 귀무가설에 반대되는 주장
- 연구자가 증명하고자 하는 가설
- 대립가설이 참이라는 증거를 찾는 것이 가설 검정의 목적
3. 가설 설정의 중요성
- 가설을 정확하게 설정하는 것이 가설 검정의 첫 단계이자 가장 중요한 부분
- 잘못된 가설 설정은 연구 결과의 신뢰성을 떨어뜨릴 수 있음

### 4.1.2 양측검정과 단측검정
1. 양측검정 Two-Tailed Test
- 대립가설이 특정 방향에 한정되지 않고, 차이가 존재하는지 여부만을 검증하는 방법
- 귀무가설에 대한 대립가설이 '차이가 있다'는 포괄적인 주장인 경우에 사용
- 양측검정에서는 유의 수준α를 양쪽 꼬리에 동일하게 분배
2. 단측검정 One-Tailed Test
- 대립가설이 특정 방향을 가리키는 경우에 사용
- 대립가설이 '크다' or '작다'와 같은 특정 방향성을 갖는 경우
- 유의 수준α가 하나의 꼬리에만 집중됨
3. 검정 선택의 중요성
- 양측검정은 일반적으로 더 보수적인 접근법으로 간주
- 단측검정은 연구자가 특정 방향에 대한 강한 가정을 가질 때 유용

### 4.1.3 유의 수준 (𝛼α)과 임계값
1. 유의 수준 α
- 가설 검정에서 귀무가설을 기각할 기준을 설정하는 값
- 제1종 오류를 범할 확률
2. 임계값 Critical Value
- 가설 검정에서 검정 통계량이 귀무가설을 기각할 수 있는 경계값
- 선택한 유의 수준과 검정 방법에 따라 달라짐
- z-검정에서 α = 0.05일때, 임계값은 약 ±1.96
- 단측 검정에서는 검정 통계량이 왼쪽 or 오른쪽 임계값보다 크거나 작을때 귀무가설을 기각
- 양측 검정의 경우, 검정 통계량이 양쪽 임계값(상한과 하한) 중 어느 하나라도 초과할 때 귀무가설 기각
3. p-value와의 관계
- 귀무가성이 참이라는 가정하에, 관측된 데이터보다 더 극단적인 결과가 나올 확률
- p-value < α, 귀무가설을 기각, 대립가설 채택
    - 관측된 데이터가 귀무가설 하에서 발생할 확률이 매우 낮다는 의미
- p-value > α, 귀무가설 채택
    - 귀무가설을 기각할 충분한 근거가 없다는 의미

## 4.2 가설 검정 절차
### 4.2.1 단계별 가설 검정 과정
1. 연구 가설 설정 
2. 유의 수준 설정
3. 검정 통계량 선택
    - 데이터에 따라 적합한 검정 통계량 선택
    - 예시
        - 평균 차이 비교 : t-검정, z-검정
        - 비율 비교 : z-검정
4. 검정 통계량 계산
5. 임계갑 결정 및 검정
- 결정 규치 : 검정 통계량이 임계값을 넘어설 경우, 귀무가설 기각
6. 결과 해석

### 4.2.2 p-값 계산과 해석
1. p-value
- 귀무가설이 참이라는 가정, 관측된 데이터 또는 그보다 더 극단적인 데이터가 나올 확률
2. 해석
- p-value < α, 귀무가설을 기각
    - 데이터를 통해 귀무가설이 맞을 가능성이 매우 낮음
    - 대립가설을 채택할 강력한 근거가 있다고 해석
- p-value > α, 귀무가설 채택
    - 귀무가설을 기각할 충분한 근거가 없다고 해석

### 4.2.3 결과의 통계적 유의성 평가
1. 통계적 유의성
- 연구 결과가 단순한 우연의 산물이 아닐 가능성을 평가
- 결과가 통계적으로 유의미하다는 것은 귀무가설을 기각할 수 있음을 의미
2. 유의성 판단
- p-value < α 일때, 연구 결과는 통계적으로 유의미하다고 판단
- 검정 통계량이 임계값을 넘는 경우에도 통계적 유의성 인정
3. 유의성과 실질적 중요성
- 통계적으로 유의미한 결과가 항상 실질적으로 중요한 것은 아님
- 실제로 중요한지는 별도의 해석이 필요

## 4.3 주요 가설 검정 방법
### 4.3.1 t-test
- 평균 간이 차이를 검정하기 위한 방법
1. 단일 표본 t-검정 One Sample t-test
- 표본 평균이 알려진 모집단 평균과 유의미하게 다른지를 검정
- 예시 : 회사에서 생산하는 제품의 평균 무게가 100g일때, 새로 생산된 30개 샘플의 평균 무게가 이와 다른지 검정



2. 독립 표본 t-검정 Independent Sample t-test
- 두개의 독립적인 그룹 간의 평균 차이를 검정
- 예시 : 남학생, 여학생의 수학 성적 평균을 비교할 때 사용



3. 대응 표본 t-검정 Paired Sample t-test
- 동일한 그룹에서의 두 측정값 간의 차이를 검정
- 다이어트 전후의 체중 차이를 검정할 때 사용



### 4.3.2 z-test
- 모집단 표준편차를 알고 있을 때, 표본 평균이 모집단 평균과 다른지를 검정하는 방법
- 예시 : 특정 제품의 평균 무게가 기존의 표준과 다른지를 검정할 때 사용



### 4.3.3 카이제곱 검정 Chi-Square Test
- 범주형 데이터의 관찰 빈도와 기대 빈도 간의 차이를 검정하는 데 사용
- 예시 : 특정 유전자형 분포가 예상과 일치하는지 검정할 때 사용



### 4.3.4 분산 분석 ANOVA
- 3개 이상의 그룹 간의 평균 차이를 검정하는 방법

1. 일원 분산 분석 One-way ANOVA
- 한 요인의 3개 이상의 그룹 간의 평균 차이 검정
- 예시 : 3개의 다른 교육 프로그램이 학생들의 성적에 미치는 영향 비교



2. 이원 분산 분석 Two-way ANOVA
- 2개의 요인이 있는 경우 그룹 간 평균 차이 검정
- 예시 : 2가지 교육 방법과 2가지 학습 도구가 학생들의 성적에 미치는 영향을 동시에 비교

## 4.4 가설 검정의 고급 주제
### 4.4.1 다중 비교 문제 Multiple Comparisons
- 여러 가설 검정을 동시에 수행할 때, α이 왜곡될 수 있는 문제를 의미
- 여러 번의 검정을 통해 우연히 통계적으로 유의미한 결과가 나타날 가능성이 높아지는 현상 때문
- 예시 : 20개의 독립적인 가설을 각각 α = 0.05로 검정하면, 약 1개의 가설은 실제로 유의미하지 않아도 우연히 유의미한 것으로 나타날 수 있음
- 해결 방법
    1. 본페로니 보정 Bonferroni Correction
        - α를 수행한 검정의 개수 n으로 나눠 각 검정에 대한 유의 수준을 보정하는 방법
        - 보정 후 유의수준 α = α/n
    2. FDR ; False Discovery Rate
        - 거짓 발견률을 제어하는 방법
        - benjamini-Hochberg 절차가 자주 사용됨

### 4.4.2 비모수 검정 Non-parametric Tests
- 데이터가 정규 분포를 따르지 않는 경우 사용할 수 있는 가설 검정 방법
- 데이터의 분포에 대한 가정이 없거나 약할 때 사용
- 비모수 검정 방법
    1. 월콕슨 순위합 검정  Wilcoxon Rank-Sum Test
        - 독립된 두 그룹 간의 차이를 비교할 때 사용
    2. 크루스칼-왈리스 검정 kruskal-Wallis Test
        - 3개 이상의 그룹 간의 차이를 비교할 때 사용
    3. 스피어만의 순위 상관 Spearman's Rank Correlation
        - 두 변수 간의 비선형 상관 관계를 검정할 때 사용

### 4.4.3 베이지안 가설 검정 Bayesian Hypothesis Testing
- 전통적인 빈도주의 접근과 달리, 데이터에 대한 사전 정보를 반영하여 가설을 검정하는 방법
- 사후 확률을 계산하여 가설이 참일 가능성을 평가
- 베이즈 요인 Bayes Factor
    - 두 가설 H_1, H_0에 대해 베이즈 요인은 주어진 데이터에 H_1이 H_0보다 얼마나 더 타당한지 나타내는 척도
